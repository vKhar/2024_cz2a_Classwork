# Principle of Software Design: Modular Design
- software is made up of small managable components.
- desirable outcomes of having small manageable code compoenents are
    - Decomposition : which allows use to address a more manageable sub-problem (instead of a larger more complex problem).
    - Maintainable : It becomes easier to debug programs as each routine can be tested separately - i.e., testing and optimisation of the code is made easier
    - Re-usable code: Time is not wasted re-writing code that has already been written.
    
- In Python, these components are functions.

**How do we create a modular function ?**
- Coupling and Cohesion
    * A modular design should have Tight Cohesion of code within a function ,and Loose Coupling between functions.

- A function should have ***"one purpose"*** only

````python
# Paste Code
# Refactoring Krishna code to a repeat until loop without defining functions

while True:
  score_input = input("Enter your score: ")

  if (score_input.isnumeric() and (0 <= int(score_input) <= 100)):
    score = int(score_input) # assume that score is an integer
    break
  else:
    print("Invalid input")

for lower_bound, grade in [ (50,"Fail"), (55,"D"), (60,"C"), (70,"B") ]:
  if score < lower_bound:
    print(grade)
    break
else:
  print("A")


````

````python
## Implementation Code for a modular design:
## get_input

def get_input()->int:
    while True:
        score_input = input("Enter your score: ")
        if validate(score_input) == True: ## calling a function
            return int(score_input)
        else:
            print("Invalid, re-enter")

def validate(score_input:str)->bool:
    if (score_input.isnumeric() and (0 <= int(score_input) <= 100)):
        return True
    else:
        return False

def convert_grade(score:int)->str:
    for  lower_bound, grade in [ (50,"Fail"), (55,"D"), (60,"C"), (70,"B") ]:
        if score < lower_bound:
            return grade
    else:
        return "A"
````

````python
## Test Driver Code:

convert_grade(get_input())
````

##### Some advantages of using this modular design
- re-usability: `validate()` can be re-use in another problem
- maintainability: if the problem requires the inclusion of the grades "U", "S" instead of just "Fail", changes can be done at only 1 place and will not affect other lines of code
- Abstration allows the code to be decomposed into parts to make it more readable, we can see easily that the solution has 3 components: `get_input(), validation and convert_grade()`

## Recursive Function:A function that calls itself
<ol>
<li> Base Case: The problem size cannot be broken down further, an explicit solution is returned</li>
<li> Recursive or General Case: The solution can obtained by making use of the result by calling the same function with a smaller problem size [recursive call] </li>
<li> The problem size must eventually reach the base case </li>
</ol>

### Example 1: Find the factorial of a number, n!

- Define as the product of a positive integer n,  and all the positive integers below n.
- 0! is equal to 1

````python
import math
math.factorial(2000) #5! == 5*4*3*2*1
````

````python
##iterative Factorial function :
def iter_factorial(n:int)->int:
  ret = 1
  for i in range(1,n+1):
    ret = ret * i
  return ret
````

````python
iter_factorial(2000)
````

````python
## Recursive Factorial function
def recur_factorial(n):
    ## base case
    if n == 0:
        return 1
    else:
        return n * recur_factorial(n-1)
    ## recursive case
````

````python
recur_factorial(5) # Valid
recur_factorial(0) # boundary
recur_factorial(-10) # invalid
````

### Example 3: Finding the Greatest Common Divisor of two positive integers ####
 the greatest common divisor (GCD) of two or more integers, which are not all zero, is the largest positive integer that divides each of the integers. For two integers x, y, the greatest common divisor of x and y is denoted
$gcd(x, y)$. For example:
- the GCD of 8 and 12 is 4, that is, gcd(8, 12) = 4
- gcd(40, 15) = 5
- gcd(9000, 980) = 20
- gcd(8, 0) = 8

Write an iterative and recrsive functions for `gcd(x, y)` where x, y are positive integers.

Use the Euclid Algorithm:

```
FUNCTION gcd(dividend:INTEGER, divisor:INTEGER) RETURNS INTEGER
    WHILE divisor <> 0 DO
        remainder <- dividend MOD divisor
        dividend <- divisor
        divisor <- remainder
    ENDWHILE
    RETURN dividend
ENDFUNCTION
```

````python
def gcd(dividend, divisor):
    while divisor != 0:
        dividend, divisor = divisor, dividend % divisor
    return dividend
````

````python
## Recursive Solution
def gcd(dividend, divisor):
    if divisor == 0:## base case
        return dividend
    else:
        ## general case
        return gcd( divisor, dividend%divisor)
````

### Example 4: Traversing a nested List

````python
nested_L = [1,3,5,"hello world",
                [ 21,22,23,
                    [ 31,32,33,34,[99,99,99,"4th level end"],"3rd level end"],
                  24,25, "2nd level end"
                ],
            6,7,8,
                ["abc",26,27,"2nd level end"],
            "1st level end"
            ]
````

````python
### Infinite List
nested_L.append(nested_L)
````

````python
## Iterative solution cannot traverse n level without using an explicit stack

#print(nested_L)

def traverse(nested_L, ret=[]):
    for i in nested_L:
        try:
            if type(i) == list:
                ret = traverse(i, ret)
            else:
                ret.append(i)
        except RecursionError:
            raise Exception("Max")
        except Exception as ex:
            raise ex
    return ret
try:
    print(traverse(nested_L))
except Exception as ex:
    print(ex)
````

````python
## code Widjaya
def order(n:int, args=False):
  ##base case = 0
  if n == 0:
    print(n)
    return

  if args != True:
    order(n-1, False)
    print(n)
  else:
    print(n)
    order(n-1, True)
````

````python
## Code #Guru's code
def find_mean(n, i = 0, total = 0):
  if n == []:
    return None

  if i == len(n):
    return total/i
  else:
    total += n[i]
    return find_mean(n, i + 1, total)
````
