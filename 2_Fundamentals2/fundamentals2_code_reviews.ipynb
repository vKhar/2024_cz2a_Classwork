##### Question 5 GCD of a list of integers

```python
#5.1
def gcd_compare(a, b):
    if b == 0:
        return a
    return gcd_compare(b, a % b)

def gcd(numbers):
    if len(numbers) == 0:
        return None
    if len(numbers) == 1:
        return numbers[0]
    if len(numbers) == 2:
        return gcd_compare(numbers[0], numbers[1])
    else:
        return gcd_compare(numbers[0], gcd(numbers[1:]))
```

```python
#5.2
def gcd(numbers:list)->int:
    if len(numbers) == 1:
        return numbers[0]
    else:
        if numbers[1] == 0:
            numbers.pop(1)
        else:
            numbers[0], numbers[1] = numbers[1], numbers[0] % numbers[1]
        return gcd(numbers)
```

```python
def gcd(numbers:list, a = 0)-> int:
    if a == 0:
        a = min(numbers)

    for n in numbers:
        if n % a != 0:
            return gcd(numbers, a - 1)

    return a
```
