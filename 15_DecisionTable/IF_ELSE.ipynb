#### Example: Convert a Decision table into if-else

```python
# order_gt_50 = bool(input("Order > 50"))
# small_pack = bool(input("Is small package"))
# promo_code = bool(input("Has promo code"))


print(f"{'order>50':<12}{'small_pack':<12}{'has promo':<12}{'Charges':<10}")
for order_gt_50 in (True, False):
    for small_pack in (True, False):
        for promo_code in (True, False):
            print(f"{order_gt_50:^12}{small_pack:^12}{promo_code:^12}",end='\t')
            ## use the condense table to print the actions
            if order_gt_50 and small_pack and  promo_code :
                print("free delivery")
            elif order_gt_50 and (small_pack or promo_code ):
                print("$1 delivery fee")
            else:
                print('#5 delivery fee')


```

#### Exercise 2: Printing the full Decision Table

```python
no_dis = "0"
dis_5 = "5%"
dis_10 = "10%"
print(f"{'>30':<5}{'>200':<5}{'is_m':<5}{' Disc':<5}")
for pur_gt_30 in (True, False):
    for pur_gt_200 in (True, False):
        for is_member in (True, False):
            print(f"{pur_gt_30:<5}{pur_gt_200:<5}{is_member:<5}", end="\t")

            if pur_gt_30:
                if pur_gt_200:
                    if is_member:
                        print(dis_10)
                    else:
                        print(dis_5)
                elif is_member:
                        print(dis_5)
                else:
                    print(no_dis)
            else:
                print(no_dis)
```
