```python
## Web Client

import socket
codec = 'ISO-8859-1'
#codec = "utf-8"
s = socket.socket()
#host = "www.facebook.com" 
host = "www.google.com" #DNS
#host = "pgp.mit.edu"
#host = "kharwoh.pythonanywhere.com"
port = 80
web_ip=socket.gethostbyname(host) # DNS Query

s.connect((web_ip,port) )


send_message=f"GET / HTTP/1.1\r\n"
send_message +=f"Host: {host}:{port}\r\n"
send_message +=f"Connection: close\r\n\r\n"
s.sendall(send_message.encode())
mesg=b""
while True:
    recv_message =  s.recv(1024)
    if not recv_message:
        break
    mesg += recv_message
contents = mesg.decode(codec).split("\r\n")
http_headers=[]
html=""
for i,data in enumerate(contents):
    if r"<html".upper() in data.upper():
        http_headers=contents[:i]
        html="".join(contents[i:])
        break
f=open("response.html","w")
f.write(html)
f.close()
s.close()
```
