```python
#Task 1.1
class Student3H2GP():
    def __init__(self,student_id,h2grades,gp_grade):
        self.__student_id = student_id #[1]indicate private attribute by using __
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        #[0]implement h2_grade_table as attribute
    
    def get_student_id(self):#[1] getter and setter pair
        return self.__student_id
    def set_student_id(self,student_id:str):
        if type(student_id) == str:
            self.__student_id = student_id
        else:
            print("Incorrect datatype")
    def get_h2grades(self): #[1] getter and setter pair
        return self.__h2grades
    def set_h2grades(self,h2grades:list):
        if type(h2grades) == list:
            self.__h2grades = h2grades
        else:
            print("Incorrect datatype")
    def get_gp_grade(self):#[1] getter and setter pair
        return self.__gp_grade
    def set_gp_grade(self,gp_grade:str):
        if type(gp_grade)==str:
            self.__gp_grade = gp_grade
        else:
            print("Incorrect datatype")
    #[0.5] get_h2grade_table() function
    def get_h2grade_table(self):
        h2grade_table= {"A":20, "B":17.5, "C":15, "D":12.5, "E":10, "S":5, "U":0}
        return h2grade_table
    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        total = 0
        for grade in self.__h2grades:
            total += self.get_h2grade_table()[grade]
        
        gpscore = (self.get_h2grade_table()[self.__gp_grade])/2
        total += gpscore
        return total
    def __repr__(self):#[1]return correct string
        return f"{self.__student_id}, 3H2: {self.__h2grades}, GP: {self.__gp_grade}"
#[1]initialise AND print
studentobj = Student3H2GP("1001",["A","B","C"],"B")
print(repr(studentobj))
print(studentobj.compute_uas())
```
