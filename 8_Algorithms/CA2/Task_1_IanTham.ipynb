```python
#Task 1.1
class Student3H2GP:
    def __init__(self, student_id, h2grades, gp_grade): #initialise class attributes
        #[1]indicate private attribute by using __
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        #[0] did not implement h2_grade_table as attribute

    #[1]return correct string
    def __repr__(self):
        return f"{self.__student_id}, 3H2: {self.__h2grades}, GP: {self.__gp_grade}"

    #getters
    def get_student_id(self): #[3] getter and setter pairs
        return self.__student_id
    def get_h2grades(self):
        return self.__h2grades
    def get_gp_grade(self):
        return self.__gp_grade
    # [0.5] get_h2grade_table() function
    def get_h2grade_table(self):
        return {'A': 20,
               'B': 17.5,
               'C': 15,
               'D': 12.5,
               'E': 10,
               'S': 5,
               'U': 0}

    #setters
    def set_student_id(self, new_id):
        self.__student_id = new_id
    def set_h2grades(self, new_grades):
        self.__h2grades = new_grades
    def get_gp_grade(self, new_grade):
        self.__gp_grade = new_grade

    #other methods
    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        uas = 0
        rank_points = self.get_h2grade_table()
        for h2_grade in self.__h2grades:
            for rp_grade in rank_points:
                if h2_grade == rp_grade:
                    uas += rank_points[rp_grade]
        for rp_grade in rank_points:
            if self.__gp_grade == rp_grade:
                uas += rank_points[rp_grade]/2
        return uas

#[1]initialise AND print
#printing given object
student = Student3H2GP('1001', ['A','B','C'], 'B')
print(student)
```

```python
#Task 1.2
class Student3H2GP_MT(Student3H2GP): #inherit base class and methods
    def __init__(self, student_id, h2grades, gp_grade, mt_grade): #initialise class attributes
        #[0] did not call super() to initialise
        #[1]indicate private attribute by using __
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        self.__mt_grade = mt_grade

    #[0] did not use super().__repr__()
    def __repr__(self): #overriden method
        return f"{self.__student_id}, 3H2: {self.__h2grades}, GP: {self.__gp_grade}, MT: {self.__mt_grade}"

    #getters
    def get_mt_grade(self):#[1]getter and setter
        return self.__mt_grade

    #setters
    def set_mt_grade(self, new_grade):
        self.__mt_grade = new_grade

    #other methods
    def compute_uas(self): #polymorphism, override same method in base class
        #[0] did not call super() to access parent's compute_uas and h2gradetable
        uas = 0
        rank_points = self.get_h2grade_table()
        for h2_grade in self.__h2grades:
            for rp_grade in rank_points:
                if h2_grade == rp_grade:
                    uas += rank_points[rp_grade]
        for rp_grade in rank_points:
            if self.__gp_grade == rp_grade:
                uas += rank_points[rp_grade]/2
            if self.__mt_grade == rp_grade:
                uas += rank_points[rp_grade]/2
        #[1]rebase
        uas = uas * 7/8
        return uas

#testing given object
#[1] initialize and print object
student_mt = Student3H2GP_MT('1001', ['A','B','C'], 'B', 'C')
#[1]test compute_uas
points = student_mt.compute_uas()
print(points)

#printing given object
print(student_mt)
```
