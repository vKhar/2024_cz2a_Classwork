```python
# task 1.1

class Student3H2GP():
    def __init__(self, student_id, h2_grades: list, gp_grade: str):
        #[0] did not indicate private attribute by using __
        self.student_id = student_id
        self.h2_grades = h2_grades
        self.gp_grade = gp_grade
        #[1]implement h2_grade_table as attribute and get_h2grade_table() function
        self.h2grade_table = {"A": 20, "B": 17.5, "C": 15, "D": 12.5, "E": 10, "S": 5, "U": 0}

    def get_student_id(self): #[3] getter and setter pairs
        return self.student_id

    def set_student_id(self, student_id):
        self.student_id = student_id
        return

    def get_h2grades(self):
        return self.h2_grades

    def set_h2grades(self, h2grades):
        self.h2_grades = h2grades
        return

    def get_gp_grade(self):
        return self.gp_grade

    def set_gp_grade(self, gp_grade):
        self.gp_grade = gp_grade
        return

    def get_h2_grade_table(self):
        return self.h2grade_table

    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        self.uas = 0
        for i in self.h2_grades:
            self.uas = self.uas + self.h2grade_table[i.upper()]

        self.uas = self.uas + (self.h2grade_table[self.gp_grade.upper()]) / 2

        return self.uas

    #[0]although return correct string, name is wrong. should be __repr__
    def __repre__(self):
        return f"<{self.student_id}>, 3H2: <{self.h2_grades}>, GP: <{self.gp_grade}>"

#[1]initialised AND print object
student = Student3H2GP('1001', ['A', 'B', 'C'], 'B')
print(student.__repre__())
print(student)
```

```python
# task 1.2
class Student2H2GP_MT(Student3H2GP):
    def __init__(self, student_id, h2_grades: list, gp_grade: str, mt_grade):
        #[1]call super() to initialise
        super().__init__(student_id, h2_grades, gp_grade)
        #[0] did not indicate private attribute by using __
        self.mt_grade = mt_grade

    def get_mt_grade(self): #[1]getter and setter
        return self.mt_grade

    def set_mt_grade(self, mt_grade):
        self.mt_grade = mt_grade.upper()
        return

    def compute_uas(self):
        #[2] call super() to access parent's compute_uas and h2gradetable
        self.uas = super().compute_uas()
        self.uas = self.uas + (self.h2grade_table[self.mt_grade]) / 2
        #[1]rebase
        return self.uas * (7 / 8)

    #[0] did not use super().__repr__()
    def __repre__(self):
        return f"<{self.student_id}>, 3H2: <{self.h2_grades}>, GP: <{self.gp_grade}>, MT: <{self.mt_grade}>"

#[1] initialize and print object
s = Student2H2GP_MT('1001', ['A', 'B', 'C'], 'B', 'C')
#[1]test compute_uas
print(s.compute_uas())
print(s)
```
