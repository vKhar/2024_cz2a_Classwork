```python
#Task 1.1
class Student3H2GP():
    def __init__(self, student_id, h2grades, gp_grade):
        #[0] did not indicate private attribute by using __
        self.student_id = student_id
        self.h2grades = h2grades
        self.gp_grade = gp_grade
        #[1]dictionary and get_h2grade_table() function
        self.h2grade_table = {'A': 20, 'B': 17.5, 'C': 15, 'D': 12.5, 'E': 10, 'S': 5, 'U': 0}
    def get_student_id(self): #[1] getter and setter pair
        return self.student_id
    def set_student_id(self, id):
        self.student_id = id
    def get_h2grades(self):#[1] getter and setter pair
        return self.h2grades
    def set_h2grades(self, grades):
        self.h2grades = grades
    def get_gp_grade(self):#[1] getter and setter pair
        return self.gp_grade
    def set_gp_grade(self, grade):
        self.gp_grade = grade
    def get_h2grade_table(self):
        return self.h2grade_table
    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        grade = 0
        for i in self.h2grades:
            grade = grade + self.h2grade_table[i]
        grade = grade + (self.h2grade_table[self.gp_grade]/2)
        return grade
    def __repr__(self):#[1]return correct string
        return f'student_id: {self.student_id}, 3H2: {self.h2grades}, GP: {self.gp_grade}'

#[1]initialise AND print
Student = Student3H2GP('1001', ['A', 'B', 'C'], 'B')
print(Student)
```
