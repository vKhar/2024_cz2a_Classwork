```python
#Task 1.1
class Student3H2GP():
    # [0.5] implement h2_grade_table as attribute but 
    # should be object attribute according to UML diagram 
    h2grade_table = {'A':20, 'B':17.5, 'C':15, 'D':12.5, 'E':10, 'S':5, 'U':0}
    
    def __init__(self, student_id, h2grades, gp_grade):
        #[1]indicate private attribute by using __
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        

    @property
    def student_id(self):#[3] getter and setter pair
        return self.__student_id
    @student_id.setter
    def set_student_id(self, student_id):
        self.__student_id = student_id

    @property
    def h2grades(self):
        return self.__h2grades
    @h2grades.setter
    def set_h2grades(self, h2grades):
        self.__h2grades = h2grades

    @property
    def gp_grade(self):
        return self.__gp_grade
    @gp_grade.setter
    def set_gp_grade(self, gp_grade):
        self.__gp_grade = gp_grade
    #[0] did not implement get_h2grade_table() function

    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        uas = 0
        for grade in self.h2grades:
            uas += Student3H2GP.h2grade_table[grade]
        uas += Student3H2GP.h2grade_table[self.gp_grade]/2
        return uas
    
    #[1]return correct string
    def __repr__(self):
        return f"{self.student_id}, 3H2: {self.h2grades}, GP: {self.gp_grade}"
        
```

```python
class Student3H2GP_MT(Student3H2GP):
    def __init__(self, student_id, h2grades, gp_grade, mt_grade):
        #[1]call super() to initialise
        super().__init__(student_id, h2grades, gp_grade)
        #[1]indicate private attribute by using __
        self.__mt_grade = mt_grade

    @property
    def mt_grade(self):#[1]getter and setter
        return self.__mt_grade
    @mt_grade.setter
    def set_mt_grade(self, mt_grade):
        self.__mt_grade = mt_grade
    
    def compute_uas(self):
        #[2] call super() to access parent's compute_uas and h2gradetable
        uas = super().compute_uas()
        uas += Student3H2GP_MT.h2grade_table[self.mt_grade]/2
        #[1]rebase
        return uas*7/8
    
    def __repr__(self):#[0] did not use super().__repr__()
        return f"{self.student_id}, 3H2: {self.h2grades}, GP: {self.gp_grade}, MT: {self.mt_grade}"
    
```
