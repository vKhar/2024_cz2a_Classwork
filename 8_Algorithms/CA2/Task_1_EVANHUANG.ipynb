```python
# Task 1.1

class Student3H2GP():
    def __init__(self, student_id: str, h2grades: list, gp_grade: str):
        self.__student_id = student_id #[1]indicate private attribute by using __
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        #[1]dictionary and get_h2grade_table() function
        self.__h2grade_table = {"A": 20, "B": 17.5, "C": 15, "D": 12.5, "E": 10, "S": 5, "U": 0}
    
    
    def get_student_id(self) -> str:#[1] getter and setter pair
        return self.__student_id
    
    
    def set_student_id(self, student_id: str):
        self.__student_id = student_id
    
    
    def get_h2grades(self) -> list:#[1] getter and setter pair
        return self.__h2grades
    
    
    def set_h2grades(self, h2grades: list):
        self.__h2grades = h2grades
    
    
    def get_gp_grade(self) -> str:#[1] getter and setter pair
        return self.__gp_grade
    
    
    def set_gp_grade(self, gp_grade: str):
        self.__gp_grade = gp_grade
    
    
    def get_h2grade_table(self) -> dict:
        return self.__h2grade_table
    
    
    def compute_uas(self) -> float:
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        return sum([ self.__h2grade_table[x] for x in self.__h2grades]) + 0.5 * self.__h2grade_table[self.__gp_grade] 
    
        
    def __repr__(self) -> str:#[1]return correct string
        return f"student_id: '{self.__student_id}', 3H2: {self.__h2grades}, GP: '{self.__gp_grade}'"
    

#[1]initialise AND print    
studentH2 = Student3H2GP("1001", ["A", "B", "C"], "B")
#print(x.compute_uas(),
#     x.get_student_id(),
#     x.get_h2grades(),
#     x.get_gp_grade(),
#     x.get_h2grade_table(),
#     sep="\n")
print(studentH2)
```

```python
# Task 1.2

class Student3H2GP_MT(Student3H2GP):
    def __init__(self, student_id: str, h2grades: list, gp_grade: str, mt_grade: str):
        #[1]call super() to initialise
        super().__init__(student_id, h2grades, gp_grade)
        #[1]indicate private attribute by using __
        self.__mt_grade = mt_grade
        
        
    def get_mt_grade(self) -> str:#[1]getter and setter
        return self.__mt_grade
    
    
    def set_mt_grade(self, mt_grade: str):
        self.__mt_grade = mt_grade
        
        
    def compute_uas(self) -> float:
        #[2] call super() to access parent's compute_uas and h2gradetable
        #[1]rebase
        return (7/8) * (super().compute_uas() + 0.5 * self.get_h2grade_table()[self.__mt_grade])
    
    
    def __repr__(self) -> str:#[1] use super().__repr__()
        return super().__repr__() + f", MT: '{self.__mt_grade}'"

        
#[1] initialize and print        
studentMT = Student3H2GP_MT("1001", ["A", "B", "C"], "B", "C")
#print(x.compute_uas(),
#     x.get_student_id(),
#     x.get_h2grades(),
#     x.get_mt_grade(),
#     x.get_h2grade_table(),
#     sep="\n")
#[1]test compute_uas
print(studentMT.compute_uas())
print(studentMT)
```
