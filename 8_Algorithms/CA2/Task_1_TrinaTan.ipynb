```python
#task 1.1
class Student3H2GP:
    def __init__(self, student_id, h2_grades, gp_grade):
        #[0] should indicate private attribute by using __ instead of _
        self._student_id = student_id
        self._h2_grades = h2_grades
        self._gp_grade = gp_grade
        #[1]implement h2_grade_table as attribute and get_h2grade_table() function
        self._h2grade_table = {'A': 20, 'B': 17.5, 'C': 15, 'D': 12.5, 'E': 10, 'S': 5, 'U': 0}

    #[1]return correct string
    def __repr__(self):
        return f'<{self._student_id}>, 3H2: <{self._h2_grades}>, GP: <{self._gp_grade}>'
    
    #[3] getter and setter pairs, we normally simplify getters so that they are 
    # easier to use in the subclasses
    #getters
    def get_student_id(self):
        return f'student id: {self._student_id}'
    
    def get_h2grades(self):
        return f'h2grades: {self._h2_grades}'
    
    def get_gp_grade(self):
        return f'gp_grade: {self._gp_grade}'
    
    def get_h2grade_table():
        return f'h2grade table: {self._h2grade_table}'
    
    #setters
    def set_student_id(self, studentid):
        self._student_id = studentid
        
    def set_h2grades(self, h2grades):
        self._h2_grades = h2grades
        
    def set_gp_grade(self, gpgrade):
        self._gp_grade = gpgrade
    
    #others
    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        sum_grades = 0
        for i in self._h2_grades:
            sum_grades += self._h2grade_table[i]
            
        return sum_grades + (self._h2grade_table[self._gp_grade]/2)
```

```python
#[1]initialised AND print object
student1 = Student3H2GP('1001',['A','B','C'],'B')
print(student1.compute_uas())
print(student1)
```

```python
#task 1.2
class Student3H2GP_MT(Student3H2GP):
    def __init__(self, student_id, h2_grades, gp_grade, mt_grade):
        #[1]call super() to initialise
        super().__init__(student_id, h2_grades, gp_grade)
        #[0] should indicate private attribute by using __ instead of _
        self._mt_grade = mt_grade

    #[0] did not use super().__repr__()    
    def __repr__(self):
        return f'<{self._student_id}>, 3H2: <{self._h2_grades}>, GP: <{self._gp_grade}>, MT: <{self._mt_grade}>'
    
    #getters
    def get_mt_grade(self):#[1]getter and setter pair
        return f'mt_grade: {self._mt_grade}'
    
    #setters
    def set_mt_grade(self, mtgrade):
        self._mt_grade = mtgrade
     
    #others
    def compute_uas(self):
        #[0] did not call super() to access parent's compute_uas and h2gradetable
        sum_grades = 0
        for i in self._h2_grades:
            sum_grades += self._h2grade_table[i]
        sum_grades += (self._h2grade_table[self._gp_grade]/2) + (self._h2grade_table[self._mt_grade]/2)
        #[1]rebase
        return (7/8) * sum_grades
            
```

```python
#task 1.3.2
class Student3H2GP_MT_H1content(Student3H2GP):
    def __init__(self, student_id, h2_grades, gp_grade, mt_grade, h1content_grade):
        #[1]call super() to initialise
        super().__init__(student_id, h2_grades, gp_grade)
        self._mt_grade = mt_grade
        self._h1content_grade = h1content_grade

    #[0] did not use super().__repr__()    
    def __repr__(self):
        return f'<{self._student_id}>, 3H2: <{self._h2_grades}>, GP: <{self._gp_grade}>, MT: <{self._mt_grade}>, H1 Content: <{self._h1content_grade}>'
    
    #getters
    def get_mt_grade(self):#[2]getter and setter pairs
        return f'mt_grade: {self._mt_grade}'
    
    def get_h1content_grade(self):
        return f'h1content_grade: {self._h1content_grade}'
    
    #setters
    def set_mt_grade(self, mtgrade):
        self._mt_grade = mtgrade
        
    def set_h1content_grade(h1contentgrade):
        self._h1content_grade = h1contentgrade
    
    #others
    def compute_uas(self):
        #[0] did not call super() to access parent's compute_uas and h2gradetable and rebase
        sum_grades = 0
        for i in self._h2_grades:
            sum_grades += self._h2grade_table[i]
        sum_grades += (self._h2grade_table[self._gp_grade]/2) + (self._h2grade_table[self._h1content_grade]/2) + (self._h2grade_table[self._mt_grade]/2)
        return (7/9) * sum_grades
```

```python
#task 1.4
class Student:
    #[1] correct constructor and initialisation
    def __init__(self, student_id, h2_grades, gp_grade, mt_grade, h1content_grade):
        self.__student_id = student_id
        self.__h2_grades = h2_grades
        self.__gp_grade = gp_grade
        self.__mt_grade = mt_grade
        self.__h1content_grade = h1content_grade
        self.__h2grade_table = {'A': 20, 'B': 17.5, 'C': 15, 'D': 12.5, 'E': 10, 'S': 5, 'U': 0}
        
    def compute_max_uas(self):
        #[1]initialise any 2, [1] add one more mark if all 4 types of students initialised
        sum3h2gp = Student3H2GP(self.__student_id, self.__h2_grades, self.__gp_grade).compute_uas()
        sum3h2gpmt = Student3H2GP_MT(self.__student_id, self.__h2_grades, self.__gp_grade, self.__mt_grade).compute_uas()
        sum3h2gph1 = Student3H2GP_H1Content(self.__student_id, self.__h2_grades, self.__gp_grade, self.__h1content_grade).compute_uas()
        sum3h2gpmth1 = Student3H2GP_MT_H1content(self.__student_id, self.__h2_grades, self.__gp_grade, self.__mt_grade, self.__h1content_grade).compute_uas()
        #[1]return max of 4 scores
        return sorted([sum3h2gp, sum3h2gpmt, sum3h2gph1, sum3h2gpmth1])[-1]
```
