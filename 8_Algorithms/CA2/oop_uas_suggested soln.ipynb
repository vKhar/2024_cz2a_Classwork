```python
 class Student3H2GP:
    def __init__(self, student_id: str, h2grades:list, gp_grade:str):
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        self.__h2grade_table = {'A':20, 'B': 17.5, 'C': 15, 'D': 12.5, 'E': 10, 'S': 5, 'U': 0}

    def get_student_id(self):
        return self.__student_id

    def set_student_id(self, id):
        self.__student_id = id

    def get_h2grades(self):
        return self.__h2grades

    def set_h2grades(self, h2grades):
        self.__h2grades = h2grades

    def get_gp_grade(self):
        return self.__gp_grade

    def set_gp_grade(self, gp_grade):
        self.__gp_grade = gp_grade

    def get_h2grade_table(self):
        return self.__h2grade_table

    def compute_uas(self):
        uas_gp = 0

        for grade in self.__h2grades:
            uas_gp += self.__h2grade_table[grade]
        uas_gp += self.__h2grade_table[self.__gp_grade]*0.5
        return uas_gp

    def __repr__(self):
        return f'{self.__student_id}, 3H2: {self.__h2grades}, GP: {self.__gp_grade}'
```

```python
s1 = Student3H2GP('1001', ['A', 'B', 'C'], 'B')
print(s1)
print((s1.compute_uas()))
```

```python
class Student3H2GP_MT(Student3H2GP):
    def __init__(self, student_id:str, h2grades:list, gp_grade:str, mt_grade:str):
        super().__init__(student_id, h2grades, gp_grade)
        self.__mt_grade = mt_grade

    def get_mt_grade(self):
        return self.__mt_grade

    def set_mt_grade(self, mt_grade):
        self.__mt_grade = mt_grade

    def compute_uas(self):
        uas_gp_mt_based80 = super().compute_uas() + super().get_h2grade_table()[self.__mt_grade]*0.5
        uas_gp_mt = uas_gp_mt_based80 * 7/8
        return uas_gp_mt

    def __repr__(self):
        return super().__repr__() + f', MT: {self.__mt_grade}'
```

```python
s2 = Student3H2GP_MT('1001', ['A', 'B', 'C'], 'B', 'C')
print(s2)
print(s2.compute_uas())
```

```python
class Student3H2GP_H1content(Student3H2GP):
    def __init__(self, student_id:str, h2grades:list, gp_grade:str, h1content_grade:str):
        super().__init__(student_id, h2grades, gp_grade)
        self.__h1content_grade = h1content_grade

    def get_h1content_grade(self):
        return self.__h1content_grade

    def set_h1content_grade(self, h1content_grade):
        self.__h1content_grade = h1content_grade

    def compute_uas(self):
        uas_gp_h1content_based80 = super().compute_uas() + super().get_h2grade_table()[self.__h1content_grade]*0.5
        uas_gp_h1content = uas_gp_h1content_based80 * 7/8
        return uas_gp_h1content

    def __repr__(self):
        return super().__repr__() + f', H1 Content: {self.__h1content_grade}'
```

```python
s3 = Student3H2GP_H1content('1001', ['A', 'B', 'C'], 'B', 'A')
print(s3)
print(s3.compute_uas())
```

```python
class Student3H2GP_MT_H1content(Student3H2GP):
    def __init__(self, student_id:str, h2grades:list, gp_grade:str, mt_grade: str, h1content_grade:str):
        super().__init__(student_id, h2grades, gp_grade)
        self.__mt_grade = mt_grade
        self.__h1content_grade = h1content_grade

    def get_mt_grade(self):
        return self.__mt_grade

    def set_mt_grade(self, mt_grade):
        self.__mt_grade = mt_grade

    def get_h1content_grade(self):
        return self.__h1content_grade

    def set_h1content_grade(self, h1content_grade):
        self.__h1content_grade = h1content_grade

    def compute_uas(self):
        uas_gp_mt_h1content_based90 = super().compute_uas() + super().get_h2grade_table()[self.__mt_grade]*0.5 \
                                    + super().get_h2grade_table()[self.__h1content_grade]*0.5
        uas_gp_mt_h1content = uas_gp_mt_h1content_based90 * 7/9
        return uas_gp_mt_h1content

    def __repr__(self):
        return super().__repr__() + f', MT: {self.__mt_grade}, H1 Content: {self.__h1content_grade}'
```

```python
s4 = Student3H2GP_MT_H1content('1001', ['A', 'B', 'C'], 'B', 'C', 'A')
print(s4)
print(s4.compute_uas())
```

```python
class Student:
    def __init__(self, student_id, h2grades, gp_grade, mt_grade, h1content_grade):
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        self.__mt_grade = mt_grade
        self.__h1content_grade = h1content_grade

    def compute_max_uas(self):
        result3h2gp = Student3H2GP(self.__student_id, self.__h2grades, self.__gp_grade).compute_uas()
        result3h2gp_mt = Student3H2GP_MT(self.__student_id, self.__h2grades, self.__gp_grade, self.__mt_grade).compute_uas()
        result3h2gp_h1content = Student3H2GP_H1content(self.__student_id, self.__h2grades, self.__gp_grade, self.__h1content_grade).compute_uas()
        result3h2gp_mt_h1content = Student3H2GP_MT_H1content(self.__student_id, self.__h2grades, self.__gp_grade, self.__mt_grade, self.__h1content_grade).compute_uas()
        return max(result3h2gp, result3h2gp_mt, result3h2gp_h1content, result3h2gp_mt_h1content)
```

```python
stu1 = Student('1001', ['A', 'B', 'C'], 'B', 'C', 'A')
print(stu1.compute_max_uas())
```
