```python
#Task 1.1
#assuming that the correct data types will be entered
#for initialisation and for setting

class Student3H2GP():
    def __init__(self, student_id, h2grades, gp_grade):
        #[1]indicate private attribute by using __
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        #[1]implement h2_grade_table as attribute and get_h2grade_table() function
        self.__h2grade_table = {"A": 20, "B": 17.5, "C": 15, "D": 12.5, "E": 10, "S": 5, "U": 0}

    #[3] getter and setter pairs
    def get_student_id(self):
        return self.__student_id

    def set_student_id(self, student_id):
        self.__student_id = student_id

    def get_h2grades(self):
        return self.__h2grades

    def set_h2grades(self, h2grades):
        self.__h2grades = h2grades

    def get_gp_grade(self):
        return self.__gp_grade

    def set_gp_grade(self, gp_grade):
        self.__gp_grade = gp_grade

    def get_h2grade_table(self):
        return self.__h2grade_table

    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        total_grade = 0
        for grade in self.__h2grades:
            total_grade += self.__h2grade_table[grade]
        total_grade += self.__h2grade_table[self.get_gp_grade()] * 0.5
        return total_grade

    #[1]return correct string
    def __repr__(self):
        return f"<{self.__student_id}>, 3H2: <{self.__h2grades}>, GP: <{self.__gp_grade}>"

#[1]initialised AND print object
StudentA = Student3H2GP("1001", ["A", "B", "C"], 'B')
print(StudentA)
```

```python
#Task 1.2
#assuming that the correct data types will be entered
#for initialisation and for setting
class Student3H2GP_MT(Student3H2GP):
    def __init__(self, student_id, h2grades, gp_grade, mt_grade):
        #[1]call super() to initialise
        super().__init__(student_id, h2grades, gp_grade)
        #[1]indicate private attribute by using __
        self.__mt_grade = mt_grade

    #[1]getter and setter
    def get_mt_grade(self):
        return self.__mt_grade

    def set_mt_grade(self, mt_grade):
        self.__mt_grade = mt_grade

    def compute_uas(self):
        #[1] call super() to access parent's h2gradetable but not compute_uas
        total_grade = 0
        grades_table = super().get_h2grade_table()
        grades = super().get_h2grades()
        for grade in grades:
            total_grade += grades_table[grade]
        total_grade += grades_table[self.get_gp_grade()] * 0.5
        total_grade += grades_table[self.get_mt_grade()] * 0.5
        #[1]rebase
        return total_grade * 7/8

    def __repr__(self):
        #[0] did not use super().__repr__()
        gp_grade = super().get_gp_grade()
        student_id = super().get_student_id()
        h2grades = super().get_h2grades()
        return f"<{student_id}>, 3H2: <{h2grades}>, GP: <{gp_grade}>, MT: <{self.__mt_grade}>"

#[1] initialize and print object
StudentB = Student3H2GP_MT("1001", ["A", "B", "C"], 'B', "C")
#[1]test compute_uas
print(StudentB.compute_uas())
print(StudentB)
```

```python
#Task 1.4
#assuming that the correct data types will be entered
#for initialisation and for setting

class Student():
    #[1] correct constructor and initialisation
    def __init__(self, student_id, h2grades, gp_grade, mt_grade, h1content_grade):
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        self.__mt_grade = mt_grade
        self.__h1content_grade = h1content_grade

    def compute_max_uas(self):
        #[1]initialise any 2, [1] add one more mark if all 4 types of students initialised
        #[1]return max of 4 scores
        max_grade = 0
        student_id = self.__student_id
        h2grades = self.__h2grades
        gp_grade = self.__gp_grade
        dummy_student = Student3H2GP(student_id, h2grades, gp_grade)
        potential_grade = dummy_student.compute_uas()
        if potential_grade > max_grade:
            max_grade = potential_grade
        dummy_student = Student3H2GP_MT(student_id, h2grades, gp_grade, self.__mt_grade)
        potential_grade = dummy_student.compute_uas()
        if potential_grade > max_grade:
            max_grade = potential_grade
        dummy_student = Student3H2GP_H1content(student_id, h2grades, gp_grade, self.__h1content_grade)
        potential_grade = dummy_student.compute_uas()
        if potential_grade > max_grade:
            max_grade = potential_grade
        dummy_student = Student3H2GP_MT_H1content(student_id, h2grades, gp_grade, self.__mt_grade, self.__h1content_grade)
        potential_grade = dummy_student.compute_uas()
        if potential_grade > max_grade:
            max_grade = potential_grade
        return max_grade

#[1]instantiate Student Object
StudentZ = Student("1001", ["A", "B", "C"], 'B', "C", "A")
#[1] compute max UAS correctly
print(StudentZ.compute_max_uas())
```
