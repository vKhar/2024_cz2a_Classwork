```python
#Task 1.1
#assume all inputs are valid
class Student3H2GP():
    def __init__(self,student_id,h2grades,gp_grade):
        #[1]indicate private attribute by using __
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        #[1]implement h2_grade_table as attribute and get_h2grade_table() function
        self.__h2grade_table = {"A":20,"B":17.5,"C":15,"D":12.5,"E":10,"S":5,"U":0}

    #[3] getter and setter pairs
    def get_student_id(self):
        return self.__student_id

    def set_student_id(self,student_id):
         self.__student_id = student_id

    def get_h2grades(self):
        return self.__h2grades

    def set_h2grades(self,h2grades):
        self.__h2grades = h2grades

    def get_gp_grade(self):
        return self.__gp_grade

    def set_gp_grade(self,gp_grade):
        self.__gp_grade = gp_grade

    def get_h2grade_table(self):
        return self.__h2grade_table

    def compute_uas(self):
        #[1]access self.__h2 grade table to compute UAS
        #[1]correct computation of UAS
        uas = 0
        for grade in self.get_h2grades():
            uas += self.get_h2grade_table()[grade]
        return uas + self.get_h2grade_table()[self.get_gp_grade()]*0.5

    def __repr__(self):#[1]return correct string.
        #can access the attributes themselves without the getter inside the class
        return f"{self.get_student_id()}, 3H2: {self.get_h2grades()}, GP: {self.get_gp_grade()}"

```

```python
#task 1.2
#assume all inputs are valid

class Student3H2GP_MT(Student3H2GP):
    def __init__(self,student_id,h2grades,gp_grade,mt_grade):
        #[1]call super() to initialise
        super().__init__(student_id,h2grades,gp_grade)
        #[1]indicate private attribute by using __
        self.__mt_grade = mt_grade

    def get_mt_grade(self):
        #[1]getter and setter pair
        return self.__mt_grade

    def set_mt_grade(self,mt_grade):
        self.__mt_grade = mt_grade

    def compute_uas(self):
        #[2] call super() to access parent's compute_uas and h2gradetable
        #[1]rebase
        return (super().compute_uas() + super().get_h2grade_table()[self.get_mt_grade()]*0.5)*7/8

    def __repr__(self): #[1] use super().__repr__()
        return super().__repr__() + f', MT: {self.get_mt_grade()}'
```

```python
#Task 1.4
#assume all inputs are valid

class Student():
    #[1] correct constructor and initialisation
    def __init__(self,student_id,h2grades,gp_grade,mt_grade,h1content_grade):
        self.__student_id = student_id
        self.__h2grades = h2grades
        self.__gp_grade = gp_grade
        self.__mt_grade = mt_grade
        self.__h1content_grade = h1content_grade

    def compute_max_uas(self):
        student_id = self.__student_id
        h2grades = self.__h2grades
        gp_grade = self.__gp_grade
        mt_grade = self.__mt_grade
        h1content_grade = self.__h1content_grade
        #[1]initialise any 2, [1] add one more mark if all 4 types of students initialised
        #[1]return max of 4 scores
        return max(Student3H2GP(student_id,h2grades,gp_grade).compute_uas()
                   ,Student3H2GP_MT(student_id,h2grades,gp_grade,mt_grade).compute_uas()
                   ,Student3H2GP_H1content(student_id,h2grades,gp_grade,h1content_grade).compute_uas()
                   ,Student3H2GP_MT_H1content(student_id,h2grades,gp_grade,mt_grade,h1content_grade).compute_uas())
```
