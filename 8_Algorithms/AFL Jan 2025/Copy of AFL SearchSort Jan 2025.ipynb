## Linear Search

Please implement a Linear Search function, lin_search(A:array, target:object)

Your implementation should take in 2 parameters:

    an array, A of objects, and
    a target object.

Your implementation should return the index of the first instance of the target object within the given array, or else, if not found, returns -1.

```python
def lin_search(A, target):

    for i in range(A):

        if A[i] == target:

            return i

    return -1
```

## Binary Search

Please implement a Binary Search function named bin_search(A,target).

Your implementation should take in 2 parameters:

    A,an array of sorted objects (in ascending order), and
    target, a target object.

Your implementation should return the index of the first instance of the target object within the given array, or else, if not found, returns -1.

```python
def bin_search(A,target, add = 0):

    pivot = len(A) // 2

    if target > A[pivot]:

        new_lst = A[pivot:]

        add += len(A[0:pivot])

        return bin_search(new_lst, target, add)

    elif target < A[pivot]:

        new_lst = A[0:pivot]

        return bin_search(new_lst, target)

    else:

        return pivot + add
```

```python
def bin_search(A:list, target):

    lower_bound = 0

    upper_bound = len(A)-1

    mid = (lower_bound+upper_bound)//2

    while lower_bound < upper_bound:

        if target == A[mid]:

            return mid

        elif target > A[mid]:

            lower_bound = mid + 1

        else:

            upper_bound = mid - 1

    return False
```

## Quick Sort



Please implement a Quicksort function named q_sort(A).

Your implementation should take in 1 parameter:

- an array A of objects.

Your implementation should also return an array with the contents of the given array sorted in ascending order.


```python
def q_sort(A):
    if len(A) < 2:
        return A

    pivot = A[0]

    left = [x for x in A[1:] if x <= pivot]

    right = [x for x in A[1:] if x > pivot]

    return q_sort(left) + [pivot] + q_sort(right)
```

```python
def q_sort(A):
    if len(A) < 2:
        return A

    piv = A.pop()

    l = []

    r = []

    for i in A:

        if i < piv:

            l.append(i)

        else:

            r.append(i)

    return q_sort(l) + [piv] + q_sort(r)
```

```python
def q_sort(A):

    if len(A) < 2:

        return A

    lower = []
    equal = []
    upper = []

    mid = len(A)//2

    for i in range(len(A)):

        if A[i] < A[mid]:

            lower.append(A[i])

        elif A[i] > A[mid]:

            upper.append(A[i])

        else:
            equal.append(A[i])

    return q_sort(lower) + equal + q_sort(upper)
```

## Merge Sort

Implement an iterative version of the Merge Sort algorithm named merge_sort(A) where A is a list.

> One possible Solution:
>
> 1) Convert a list of items into a L, a list of list of 1 item:
> 	e,g. [1,2,3] -> [ [1],[2],[3] ]
> 2) while len(L) > 1 do
> 		pop the first 2 lists a and b from L,
> 		c = merge(a,b) // you need to implement merge()
> 		//merge() will merge the 2 sorted list a, b into a sorted list c
> 		L.append(c)
> 3) return L[0]

```python
def merge(A,B):

    merged = []

    while A and B:

        if A[0] < B[0]:

            merged.append(A.pop(0))

        else:

            merged.append(B.pop(0))

    if A:

        merged = merged + A

    if B:

        merged = merged + B

    return merged

def merge_sort(A):

    if len(A) <= 1:

        return A



    middle = len(A) // 2

    left = A[:middle]

    right = A[middle:]

    return merge(left, right)
```

```python
def merge(A,B):

    merged = []

    while (len(A) >0 and len(B) > 0):

        if (A[0] > B[0]):

            merged.append(B.pop(0))

        else:

            merged.append(A.pop(0))

    merged = merged + A + B

    return merged

def merge_sort(A):

    if (len(A) <= 1):

        return A

    L = [[i] for i in A]

    while (len(L) > 1):

        c = merge(L.pop(0),L.pop(0))

        L.append(c)

    return L[0]
```
