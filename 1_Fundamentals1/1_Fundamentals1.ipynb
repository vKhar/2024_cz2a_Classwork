Validation :

- Password cannot be an empty string.
- At least 1 lowercase letter between `[a-z]` and 1 uppercase letter between `[A-Z]`.
- At least 1 number between `[0-9]`.

````python
password = input("Enter your password: ")

if len(password) == 0 :
    print("Write something")

elif 6 <= len(password) <=16:
      password.isupper() and password.islower() == True,
      print("Correct Password")
else:
    print("Wrong Password")
````

````python
## Aditya's submission
password = input("Enter a password:")

num = False
upper = False
lower = False
special = False

if 6<=len(password) <=16 :
    for char in password:
        if char.islower() == True:
            lower = True

        if char.isupper() == True:
            upper = True

        if char.isdigit() == True:
            num = True
        if char in ["$","#","@"]:
            special = True
    if num == True and upper == True and lower == True and special == True:
        print("Valid password")
    else:
        print("Invalid")

else:
    print("Invalid")
````

# Problem 6: Find the terms of a geometric sequence

## Geometric Sequence

In mathematics, a **geometric sequence** is a sequence of real numbers
$$u_1,u_2,\cdots, u_n$$
such that the ratio between two consecutive terms is constant, i.e. $\frac{u_{n+1}}{u_n}=r$ for some constant $r\in \mathbb{R}\backslash \{0,1\}$, called the **common ratio** and $n\geq 1$.

For example, the sequence $2,6,18,54,162,\dots$ is an arithmetic sequence with first term $2$ and common ratio $3$.


The problem:
- Request a numeric input from the user to be used as the first term.  Let us denote this input as `first_term`.
- Request a numeric input from the user to be used as the common difference.  Let us denote this input as `common_ratio`.
- Request an integer input from the user for the number of terms.  Let us denote this input as `number_of_terms`.
- Print out the sequence up to the nth term.

Example interaction:

>Please input the first term in the geometric sequence:<br>
>2<br>
>Please input the common ratio:<br>
>3<br>
>Please input the number of terms:<br>
>5<br>
>The first 5 terms of the geometric sequence are:<br>
>2<br>
>6<br>
>18<br>
>54<br>
>162<br>

````python
# Request inputs from the user
first_term = int(input("Enter the first term of the sequence: "))
common_ratio = int(input("Enter the common ratio of the sequence: "))
number_of_terms = int(input("Enter the number of terms in the sequence: "))

# Initialize an empty list to store the sequence
sequence = []

# Generate the sequence up to the nth term
term = first_term
for _ in range(number_of_terms):
    sequence.append(term)
    term *= common_ratio

# Print out the sequence
print("The sequence is:", sequence)
````

## Problem 6A: Compound Interest Problem

A friend of his that work in the company offers him another deal: same capital \$1000, same yearly interest rate of 100% but applied quarterly for a year, meaning every 3 months, the applied interest rate to the current amount is $\frac{100}{4}\%=25\%$. So, after 3rd month, his investment becomes \$1250. After 6th month, it's \$1562.50. After 9th month, it's \$1953.13. Under this scheme,

a)  How much would Mr. $B$'s investment worth after a full year?

Compound Interest formula : $P(1+interest)^{\text{num\_payments}}$

````python
from my_stuff import *
## I have a live car detection ADT named Car_Detector that can
## detect colours of car from a webcam and return the first n cars detected
## count the number of cars with the same colours.

## dynamic binning
colours_bin = {}

for colour in Car_Detector(555):
    if colour in colours_bin :
        colours_bin[colour] +=1
    else: ## bin not created yet
        colours_bin[colour] = 1
````
