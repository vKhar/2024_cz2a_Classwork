##### Modulo Arithmetic to represent circular relationship
```
["stone", "scissors", "paper"]
if p2 - p1 > 0 , it means p2 is in front of p1  and p1 takes (p2-p1) % 3  steps forward(clockwise) to reach p2
if p2 - p1 < 0 , it means p2 is behind p1 and p1 takes (p2-p1) % 3 steps forward (clockwise) to reach p2
```

See the following [Modulo Arithmetic](https://www.khanacademy.org/computing/computer-science/cryptography/modarithmetic/a/congruence-modulo) for Modulo Arithmetic

````python
## 3.3 using modulo in another form
steps = ["stone", "scissors", "paper"]
player_1 , player_2 = "stone", "scissors"
step_1 = steps.index(player_1)
step_2 = (step_1 + 1 ) % 3
#diff = step_2 - step_1

if player_1 == player_2:
    print("Tie")
elif steps[step_2] == player_2 :
    print("Player 1 wins")
else:
    print("Player 2 wins")
````
