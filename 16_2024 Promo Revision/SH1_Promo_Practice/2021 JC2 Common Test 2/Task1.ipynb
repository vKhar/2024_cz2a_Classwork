```python
def task1_1(in_array):
    look_up = {}
    for item in in_array:
        if item in look_up:
            look_up[item] += 1
        else:
            look_up[item] = 1
    return look_up
```

```python
task1_1(["Alice", "Bob", "Alice", "Charlie", "Bob", "Bob", "Alice", "Charlie", "Alice"])
```

```python
task1_1(["1", "6", "10", "6", "2", "1", "9", "1", "6", "9", "1", "2"])
```

```python
task1_1(["Yes", "Yes", "No", "Yes", "No", "No", "No", "Yes", "No", "Yes"])
```

```python
import string
def task1_2(sentence):
    return task1_1( sentence.translate({ ord(c): None for c in string.punctuation}).split())
```

```python
import string
def task1_2(sentence):
    #remove punctuations
    sentence = "".join([ c for c in sentence if c not in string.punctuation])
    return task1_1(sentence.split())
```
