```python
# Task 3.2

import csv

record = {}  # year as key, {min: ..., max: ... } as value
f = open('surface_air_temperature.csv', 'r')
for row in csv.DictReader(f):
    year, month = row['month'].split('-')
    curr_min = float(row['temp_mean_daily_min'])
    curr_max = float(row['temp_mean_daily_max'])
    curr_range = curr_max - curr_min
    if year not in record:
        record[year] = {'min': None, 'max': None, 'range': None}

    if not record[year]['min'] or curr_min < record[year]['min'][0]:
        record[year]['min'] = (curr_min, month)

    if not record[year]['max'] or curr_max > record[year]['max'][0]:
        record[year]['max'] = (curr_max, month)

    if not record[year]['range'] or curr_range > record[year]['range'][0]:
        record[year]['range'] = (curr_range, month)
f.close()

for year, rec in record.items():
    print(f"{year}: min={rec['min'][0]} (month {rec['min'][1]}), "
          f"max={rec['max'][0]} (month {rec['max'][1]}), "
          f"range={rec['range'][0]} (month {rec['range'][1]})")
```

```python
# Task 3.3

import csv

record = {}  # month as key, {year: {min: ..., max: ... }} as value
f = open('surface_air_temperature.csv', 'r')
for row in csv.DictReader(f):
    # format: YYYY-MM
    year, month = row['month'].split('-')
    curr_min = float(row['temp_mean_daily_min'])
    curr_max = float(row['temp_mean_daily_max'])
    if month not in record:
        record[month] = {}
    record[month][year] = {'min': curr_min, 'max': curr_max}
f.close()

record_avgs = {}
for month in record.keys():
    avg_min = 0
    avg_max = 0
    years = 0
    
    # Sum up avg min, avg max, and count years
    for year in record[month].keys():
        avg_min = avg_min + record[month][year]['min']
        avg_max = avg_max + record[month][year]['max']
        years = years + 1

    # Calculate averages    
    avg_min = avg_min / years
    avg_max = avg_max / years
    record_avgs[month] = {'min': avg_min, 'max': avg_max}

for month in record_avgs.keys():
    print(f"Month {month}: "
          f"average daily min 1982-2020 was {record_avgs[month]['min']}, "
          f"deviation in 2020 was {record[month]['2020']['min'] - record_avgs[month]['min']} deg C")
    print(f"Month {month}: "
          f"average daily max 1982-2020 was {record_avgs[month]['max']}, "
          f"deviation in 2020 was {record[month]['2020']['max'] - record_avgs[month]['max']} deg C")
```
