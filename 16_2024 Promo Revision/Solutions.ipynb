```python
#Task 1.1
def task1_1(in_array):
    bin = {}
    for item in in_array:
        if item in bin:
            bin[item] +=1
        else:
            bin[item] = 1
    return bin
```

```python
# Task 1.2
import string
def task1_2(sentence):
    new_sentence = ""

    # remove punctuations
    for char in sentence:
        if char not in string.punctuation:
            new_sentence += char
    words = new_sentence.split()
    return task1_1(words)
```

```python
# Task 1.3

def insert_sort(in_array):
    for boundary in range(1,len(in_array)):
       for back_ptr in range(boundary,0,-1):
           if in_array[back_ptr][1] > in_array[back_ptr-1][1]:
                in_array[back_ptr-1], in_array[back_ptr] = in_array[back_ptr], in_array[back_ptr-1]
           else:
               break     
```

```python
def bubble_sort(in_array):
    for i in range(len(in_array)-1):
        swap = False
        for j in range(len(in_array)-1-i):
            if in_array[j][1] < in_array[j+1][1]:
                in_array[j], in_array[j+1] = in_array[j+1], in_array[j]
            swap = True
        if not swap:
            break
```

```python
# Task 2.1
## IIII == IV
## VIIII = IX
## XXXXIIIII == IL
digits = {
    50 : "L",
    49 : "IL",
    10 : "X",
    9 : "IX",
    5 : "V",
    4: "IV",
    1 : "I"
}
def to_roman(n, validate=True):
    
    if validate:
        if type(n) != int or  n < 1 or n > 20:
            print("Invalid input")
            return None
        
    ret = ""

    for digit in sorted(digits.keys(), reverse=True):
        if n in digits:
            return ret + f"{digits[n]}"
        
        quotient, remainder = divmod(n, digit)
        ret += digits[digit]*quotient
        n = remainder
    return ret

```

```python

reverse_digits = dict([ reversed(item) for item in digits.items() ])
```

```python
# Task 2.3
def to_int(roman):
    ret = 0
    while len(roman) != 0:
        if roman[:2] in reverse_digits:
            ret += reverse_digits[roman[:2]]
            roman = roman[2:]
        else:    
            ret += reverse_digits[roman[0]]
            roman = roman[1:]
    return ret 
```

```python
def add_roman(r1, r2):

    i1, i2 = to_int(r1), to_int(r2)
    #print(i1,i2)
    return to_roman(i1+i2,validate=False)
    
```
