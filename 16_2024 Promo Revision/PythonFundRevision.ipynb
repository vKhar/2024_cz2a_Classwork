```python
def bin_search(A,target):
    
    def recur(lb, ub):
        if lb > ub :
            return None
        mid = (lb+ub)//2
        if target == A[mid]:
            return mid
        elif target < A[mid]:
            return recur( lb, mid-1)
        else:
            return recur( mid+1, ub)

    return recur(0, len(A)-1)
```

```python
bin_search( [2,4,6,8,10], 8)
```
