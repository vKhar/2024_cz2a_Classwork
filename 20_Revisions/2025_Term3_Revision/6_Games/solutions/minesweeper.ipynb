```python
BOMB="X"
EMPTY="_"
class Cell:
    def __init__(self,data=EMPTY):
        self.data = data
        self.switch = False # False:closed, True:opened
    def flip(self):
        self.switch = True
    
    def __str__(self):
        if self.switch:
            return f"{str(self.data):>2}"
        else:
            return f"{EMPTY:>2}"
class Grid:
    def __init__(self,n):
        self.size = n
        self.grid = [ [ Cell()for _ in range(n)] for _ in range(n) ]
    def print(self):
        print("\n")
        for row in self.grid:
            line=""
            for cell in row:
                line+=f"{cell}"
            print(line)
    def index_to_grid(self, i):
        return divmod(i,self.size)
    def grid_to_index(self, x, y):
        return (self.size)*x + y
    def fill_bombs(self,n):
        import random
        if n <= self.size ** 2:
            for index in random.sample(range(self.size**2), n):
                x, y = self.index_to_grid(index)
                self.grid[x][y].data = BOMB
    def get_neighbours(self, index): 
        neighbours = []
        x, y = self.index_to_grid(index)
        for row in range(x-1, x+2):
            for col in range(y-1, y+2):
                if (0 <= row < self.size) and  (0 <= col < self.size) and ( (row,col) != (x,y) ):
                 neighbours.append((row,col))
        return neighbours

    def fill_clues(self):
        for index in range(self.size**2):
            bomb_count = 0
            x, y = self.index_to_grid(index)
            if self.grid[x][y].data == BOMB:
                continue
            for neighbour in self.get_neighbours(index):
                xx,yx = neighbour
                if self.grid[xx][yx].data == BOMB:
                    bomb_count +=1
            self.grid[x][y].data = bomb_count
    def play(self):
        self.fill_bombs(3)
        self.fill_clues()
        while True:
            x, y = int(input("X")), int(input("Y"))
            self.grid[x][y].flip()
            self.print()
            if self.grid[x][y].data == BOMB:
                print("BOMMMMBBB")
                break
            


            
        
```
