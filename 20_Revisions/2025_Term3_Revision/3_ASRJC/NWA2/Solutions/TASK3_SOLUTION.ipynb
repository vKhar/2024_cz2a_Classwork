```python
# Task 3.1
import pymongo
import json

# connect to MongoDB server
client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("shop")
coll = db.get_collection("inventory")

# file closes after with block
with open("INVENTORY.JSON", "r") as file:
    data = json.load(file)["inventory"]  # list of dictionaries

coll.drop()  # drop collection if currently exists

# insert data into MongoDB collection
coll.insert_many(data)
```

```python
# Task 3.2
result = coll.find({"material": "Silk"}, {"_id": 0, "type": 1, "brand": 1, "price": 1}).sort("price", -1)
for item in result:
    print(item)
```

```python
# Task 3.3
result = coll.find({
    "brand": {"$in": ["Brioni", "Canali", "Brooks Brothers"]}, 
    "price": {"$gte": 200, "$lte": 2000}
    }, 
    {"_id": 0, "type": 1, "brand": 1, "material": 1, "price": 1})

for item in result:
    print(item)
```

```python
# Task 3.4
# Extract items to donate
donations = coll.find({"type": {"$in": ["jacket", "tie", "pocket_square"]}})

# Compute total value
total_value = 0
for item in donations:
    total_value += item["price"] * item["stock"]

# Output
print(f"Total value of items to donate: ${total_value}")

# Remove from inventory
coll.delete_many({"type": {"$in": ["jacket", "tie", "pocket_square"]}})

# Get remaining items
result = coll.find({})
item_ids = [item["product_id"] for item in result]
print("Remaining items:")
print(", ".join(item_ids))
```

```python
# Close connection
client.close()
print("Connection closed.")
```
