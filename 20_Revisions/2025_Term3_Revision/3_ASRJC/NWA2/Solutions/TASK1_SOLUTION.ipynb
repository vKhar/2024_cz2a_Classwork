```python
# Task 1.1
import random


rand_list = []  # initialise list to store random numbers

# ensure final list has length 20
while len(rand_list) < 20:
    num = random.randint(1, 50)
    if num not in rand_list:  # check for duplicates
        rand_list.append(num)  # add unique number

# show output
print("List:", rand_list)
print("Length:", len(rand_list))
```

```python
# Task 1.2
class BST:
    def __init__(self, data=None):
        # Constructor
        self.data = data
        self.left = None
        self.right = None
    
    def insert(self, value):
        if self.data is None:  # insert first value
            self.data = value
            return
        
        if value < self.data:  # insert left
            if self.left is None:
                self.left = BST(value)  # create new node
            else:
                self.left.insert(value)  # insert into left subtree
        else:  # insert right
            if self.right is None:
                self.right = BST(value)
            else:
                self.right.insert(value)  # insert into right subtree
    
    def search(self, value):
        if self.data is None:  # empty tree
            return False
        
        if self.data == value:  # found value
            return True
        elif value < self.data:  # search left subtree
            if self.left is None:
                return False
            return self.left.search(value)
        else:  # search right subtree
            if self.right is None:
                return False
            return self.right.search(value)
    
    def preorder(self):
        # Pre-order traversal (root, left, right)
        result = []
        if self.data is not None:
            result.append(self.data)
            if self.left is not None:
                result.extend(self.left.preorder())
            if self.right is not None:
                result.extend(self.right.preorder())
        return result
    
    def rev_inorder(self):
        # Reverse in-order traversal (right, root, left)
        result = []
        if self.data is not None:
            if self.right is not None:
                result.extend(self.right.rev_inorder())
            result.append(self.data)
            if self.left is not None:
                result.extend(self.left.rev_inorder())
        return result
    
# Create BST and insert numbers
nums = BST()
for num in rand_list:
    nums.insert(num)

# Output in pre-order
print("Pre-order Traversal:", nums.preorder())
print("Search for 12:", nums.search(12))
print("Search for 47:", nums.search(47))
```
