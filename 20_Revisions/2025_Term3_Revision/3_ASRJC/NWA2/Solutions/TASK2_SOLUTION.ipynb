```python
# Task 2.1
import sqlite3
import csv


# Create db
conn = sqlite3.connect("CTF.DB")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Participant (
    ID TEXT PRIMARY KEY,
    team TEXT,
    name TEXT,
    data_sci INTEGER,
    crypto INTEGER,
    forensics INTEGER,
    web INTEGER,
    rev_eng INTEGER,
    pwn INTEGER
)""")

# Read data from text file
# File closes after with block
with open("LEADERBOARD.TXT", "r") as file:
    for line in file:
        parts = line.strip().split(" | ")
        id_val = parts[0]
        team = parts[1]
        name = parts[2]
        scores = parts[3].split(", ")
    
        # Insert data
        cursor.execute("""
        INSERT INTO Participant(ID, team, name, data_sci, crypto, forensics, web, rev_eng, pwn)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)""",
        (id_val, team, name, scores[0], scores[1], scores[2], scores[3], scores[4], scores[5]))

conn.commit()

# Query to count records in the table
cursor.execute("SELECT COUNT(*) FROM Participant")
record_count = cursor.fetchone()[0]
print(f"Number of records: {record_count}")

# Close connection
conn.close()
```

```python
# Task 2.2
conn = sqlite3.connect("CTF.DB")
cursor = conn.cursor()

# Query
cursor.execute("""SELECT name FROM Participant
               WHERE web >= 1000 AND web <= 5000""")

results = cursor.fetchall()
print("Participants with Web scores from 1000 to 5000:")
for row in results:
    print(row[0])

# Close connection
conn.close()
```

```python
# Task 2.3
conn = sqlite3.connect("CTF.DB")
cursor = conn.cursor()

# Insert new record
cursor.execute("""
INSERT INTO Participant
(ID, team, name, data_sci, crypto, forensics, web, rev_eng, pwn)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
""", ("asrjc-14", "IKEA", "Baby Blahaj", 1500, 1200, 300, 0, 0, 0))
    
# Update pwn scores for Team i*99
cursor.execute("""
UPDATE Participant
SET pwn = pwn + 1234
WHERE team = 'i*99'
""")

# Delete record with ID 'asrjc-13'
cursor.execute("""
DELETE FROM Participant
WHERE ID = 'asrjc-13'
""")

# Commit changes
conn.commit()

# Extract data for CSV
cursor.execute("""
SELECT ID, name, (data_sci + crypto + forensics + web + rev_eng + pwn) AS overall_score
FROM Participant
ORDER BY overall_score DESC
""")

results = cursor.fetchall()

# Write to CSV
# file closes after with block
with open("PROCESSED.CSV", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["ID", "name", "overall_score"])
    for row in results:
        writer.writerow(row)

print("Data written to CSV file.")

# Close connection
conn.close()
```
