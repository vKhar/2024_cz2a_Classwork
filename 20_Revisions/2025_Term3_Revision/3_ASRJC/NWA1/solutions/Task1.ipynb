```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

    def __repr__(self):
        return f"{self.data}"
    
    def what_ever(self):
        self.data

class LinkedList:
    def __init__(self):
        self.start = None

    def insert_at_start(self, data):
        if self.start == None:
            self.start = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.start
            self.start = new_node

    def get_data(self):
        ret=[]
        cur = self.start
        while cur != None:
            ret.append(repr(cur))
            cur = cur.next
        return "->".join(ret)

    def delete(self, target):
        if self.start == None:
            return False
        if self.start.data == target:
            self.start = self.start.next
            return True
        prev = self.start
        cur = self.start.next
        while cur:
            if cur.data == target:
                prev.next = cur.next
                return True
            prev = cur
            cur = cur.next
        else:
            return False
        
    def reverse(self):
        if self.start == None:
            return
        prev = None
        cur = self.start
        while cur:
            tmp = cur.next
            cur.next = prev
            prev = cur
            cur = tmp
        self.start = prev

```
