The LinkedList class should have the following methods:

- delete(target): deletes the first node in the linked list that contains the given target value. If a node containing the target is found and deleted, return True, otherwise return False.
- reverse(): reverses the linked list in-place (which should involve adjusting the pointers in the existing structure without creating a new linked list).

Test your code appropriately.


```python
# Task 1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def delete(self, target):
        # Empty Llist
        if self.head is None:
            print("Linked list is empty lah")
            return False

        # Target at the head node
        if self.head.data == target:
            self.head = self.head.next
            print("Data has been curb stomped.")
            return True

        # Anywhere else in the Linked list
        curr = self.head
        while curr.next is not None:
            if curr.next.data == target:
                # Skip the target node.
                curr.next = curr.next.next
                print("Drugs found on data, sent for execution.")
                return True

            # Advance to next node
            curr = curr.next

        # Node not found
        print("No match for target, sending drone strike to your house.")
        return False

    def reverse(self):
        curr = self.head
        prev = None

        while curr is not None:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node

        self.head = prev
```
