```python
### Q1 (a)

con = sqlite3.connect("visits.db")
try:
    con.execute(''' 
        CREATE TABLE Timestamps(DateTime TEXT NOT NULL)
    ''')
    con.commit()
except:
    pass
con.close()
```

```python
### Q1(b,c)
app = flask.Flask(__name__)

@app.route('/')
def home():
    con = sqlite3.connect("visits.db")
    rows =con.execute('SELECT * FROM Timestamps Order By DateTime Desc').fetchall()
    con.close()
    return flask.render_template('visits.html', rows=rows)

@app.route('/add/')
def add():
    con = sqlite3.connect("visits.db")
    # timestamp = f"{str(now.year)}-{str(now.month)}-{str(now.day)},{str(now.hour)}:{str(now.minute)}" 

    con.execute('INSERT INTO Timestamps(DateTime) VALUES(?)', (dt.datetime.now().strftime("%Y-%m-%d,%H:%M:%S"),))
    con.commit()
    con.close()
    return flask.redirect(flask.url_for('home'))

@app.route('/clear/')
def clear():
    con = sqlite3.connect("visits.db")
    con.execute('DELETE FROM Timestamps')
    con.commit()
    con.close()
    return flask.redirect(flask.url_for('home'))

app.run()
```
