##### Lomuto Parititioning Algorithm

````python
##
## partition (L:array,lb:int,ub:int) returns pivot position of array
##  cur - pointer at cur index of array
##  p   - all items in array to the left of p are <= pivot, i.e. p will point at a value > pivot
## using cur to scan from lb to ub (inclusive), reposition p and swap items to meet the above requiements
## at the end of the partion, the array will be re-arrange according to the selected pivot
## if we use L[0] as the pivot, then L ==  [1,2,3,5,7] and returns 2
L = [3,7,2,5,1]
#   return [2,1,3,5,7], 2
def partition(L,lb,ub):
    ##  initialise , pivot, p, cur
    pivot = L[lb]
    cur = lb+1
    p = lb +1
    ## scan from lb to ub
    while cur <= ub:
        if L[cur] <= pivot:
            L[p], L[cur] =L[cur], L[p]
            p+=1
            cur+=1
        else:
            cur+=1
    L[lb], L[p-1] =  L[p-1], L[lb]
    return p-1
    ## return pivot positon

````

#### Hoare Partitioning Algorithm
- as in A Level 2018 paper 1, Question 2

````python
FUNCTION QuickSort(Scores)

  QuickSortHelper(Scores, 0, LENGTH(Scores) - 1)
  RETURN Scores

ENDFUNCTION



FUNCTION QuickSortHelper(Scores, First, Last)

  IF First < Last
    THEN

      SplitPoint ‚Üê Partition(Scores, First, Last)

      QuickSortHelper(Scores, First, SplitPoint - 1)

      QuickSortHelper(Scores, SplitPoint + 1, Last)

  ENDIF
  RETURN Scores

ENDFUNCTION

FUNCTION Partition(Scores, First, Last)

  PivotValue ‚Üê Scores[First]

  LeftMark ‚Üê First + 1

  RightMark ‚Üê Last


  Done ‚Üê FALSE

  WHILE (Done = FALSE)

    WHILE LeftMark <= RightMark AND Scores[LeftMark] <= PivotValue

      LeftMark ‚Üê LeftMark + 1
    ENDWHILE

    WHILE Scores[RightMark] >= PivotValue AND RightMark >= LeftMark

      RightMark ‚Üê RightMark - 1
    ENDWHILE
    IF RightMark < LeftMark
      THEN

        Done ‚Üê TRUE

    ELSE

      Temp ‚Üê Scores[LeftMark]

      Scores[LeftMark] ‚Üê Scores[RightMark]

      Scores[RightMark] ‚Üê Temp

    ENDIF

  ENDWHILE

  <swap Scores[First] with Scores[RightMark]>

  RETURN RightMark

ENDFUNCTION
````

````python
def partition(L,lb,ub):
    ## implement the pseudcode described in QUICKSORT.txt
    pivot_value = L[lb]
    left = lb +1
    right = ub

    done = False

    while not done:
        while left <= right and L[left] <= pivot_value:
            left += 1
        ## upon exit left will point to a value > pivot or l-r scan has complete

        while left <= right and L[right] >= pivot_value:
            right -= 1
        ## upon exit right will point to a value < pivot

        if left > right:
            ## all items scanned
            done = True
        else:
            L[left], L[right]  = L[right], L[left]

    L[lb], L[right] = L[right], L[lb]
    return right
````

````python
## in-place Quick Sort
import random
L = [random.randint(1,100) for _ in range(20)]
def qsort(L):
    def qsort_helper(lb,ub):

        ## base case
        if lb > ub:
            return
        else:##
            p = partition(L,lb,ub)
            qsort_helper(lb,p-1)
            qsort_helper( p+1, ub)
    qsort_helper(0, len(L)-1)
print(L)
qsort(L)
print(L)
````

````python
def merge(L,lb,mid,ub):
    left = lb #left subarray [lb:right]
    right = mid # right sub srray [right:ub+1]
    # print(L[lb:left], L[left:right], L[right:ub+1])
    while left < right and right <= ub:
        if L[left] < L[right]:
            left +=1
            # print(L[lb:left], L[left:right], L[right:ub+1])
        else:

            ## tmp<-R1
            ## shift the left array place to the right and replace L1, wuth R1
            ## [L1,L2,L3,R1,R2,R3] -> [L1,L1,L2,L3,R2,R3]
            ## replace L1 with R1  -> [R1,L1,L2,L3,R2,R3]

            # typical insertion sort shifting
            # tmp = L[right]
            # j = right
            # while j > left:
            #     L[j] = L[j-1]
            #     j-=1
            # L[left] = tmp

            ## shift using Evan's algorithm üòä
            for k in range(right, left, -1):
                L[k],L[k-1] = L[k-1], L[k]
            ##
            left+=1
            right+=1
            # print(L[lb:left], L[left:right], L[right:ub+1])
    # print(L[lb:ub+1])
    return lb
````

````python
def msort_hepler(L,lb,ub):
    if lb == ub :
        return lb

    mid = (lb + ub)//2
    return merge(L,
                    msort_hepler(L,lb,mid),
                    msort_hepler(L,mid+1, ub),
                    ub
                    )
````

````python
def merge_sort(L):
    if len(L) < 2:
        return L
    lb = 0
    ub = len(L)-1
    mid = (lb + ub )//2
    return merge(L,
                 msort_hepler(L,lb, mid),
                 msort_hepler(L,mid+1,ub),
                 ub)
````

### Buble Sort

````python
def bubble_sort(A):
    for i in range(len(A)-1):
        if A[i] > A[i+1]:
            A[i+1], A[i] = A[i], A[i+1]
            if bubble_sort(A):
                break
    return True
````

````python
import random
L = [4,2,1,3]
print(L)
## Evan's insertion sort

for i in range(1,len(L)):
    for j in range(i+1): ## find a position j, where L[i] should be at
        if L[i] < L[j]:
            break
    for k in range(i,j,-1): ## loop will not run when i == j
        L[k], L[k-1] = L[k-1],L[k]
print(L)
````
