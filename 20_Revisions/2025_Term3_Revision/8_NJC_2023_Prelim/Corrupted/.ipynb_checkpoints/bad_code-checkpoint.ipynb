```python
#Task 3.1
import random
def qsort(L):
    if len(L) < 2:
        return L
    index = random.randint(0, len(L)-1)
    pivot = L.pop(index)
    left = []
    right = []
    for i in L:
        if i[0] > pivot[0]:
            right.append(i)
        else:
            left.append(i)
    return qsort(left) + [pivot] + qsort(right)
```

```python
with open("WORDS.TXT",'r') as f:
    content = f.read().strip().split('\n')
words = [(content[i], i + 1) for i in range(len(content))]
```

```python
sorted_partial = []
while len(words) > 3000:
    to_be_sorted = words[:2999]
    words[:2999] = []
    sorted_tmp = qsort(to_be_sorted)
    sorted_partial.append(sorted_tmp)
sorted_partial.append(words)
```

```python
def merge(list1,list2):
    ret = []
    while len(list1) and len(list2):
        if list1[0] < list2[0]:
            ret.append(list1.pop(0))
        else:
            ret.append(list2.pop(0))
    print(ret)
    return ret + list1 + list2
```

```python
sorted_all = []
while len(sorted_partial) > 1:
    sorted_partial[0] = (merge(sorted_partial[0],sorted_partial[1]))
sorted_all = sorted_partial
print(sorted_all)
```
