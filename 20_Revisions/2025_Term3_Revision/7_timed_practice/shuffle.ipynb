```python
## Given a list of items you can use the shuffle function from the random module to shuffe the
## items in the list inplace
import random
# random.seed(1) # this will always give the same result
deck = [ 1,2,3,4,5]
random.shuffle(deck)
deck
```

```python
## Transpose a list of list
LoL = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
tLoL=[]
for col in range(len(LoL[0])):
    tLoL.append([ LoL[row][col] for row in range(len(LoL)) ])
tLoL
```
