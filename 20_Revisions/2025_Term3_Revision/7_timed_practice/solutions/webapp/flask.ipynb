```python

from flask import Flask, render_template, redirect, url_for,request
from Tubes import *

app = Flask(__name__)
g_tubes = initialise_game()

def tubes_list(tubes):
    c = tubes[0].capacity
    display = [[None for _ in range(c)] for _  in range(c)]
    for i in range(c):
        for j,item in enumerate(tubes[i].get_items()):
            display[j][i] = item
    return display

@app.route('/')
def home_default():
    return home("continue")

@app.route('/<status>')
def home(status):
    global g_tubes
    return render_template('home.html', tubes=tubes_list(g_tubes),status=status)

@app.route('/move', methods=['POST'])
def make_move():

    global g_tubes
    source = int(request.form['source'])
    destination = int(request.form['destination'])
    if move(g_tubes, source, destination):
        if game_won(g_tubes):
            return redirect("/won")
        else:
            return redirect("/continue")
    else:
        return redirect("/invalid")
    

@app.route('/game/<status>')
def game(status):
    return render_template('game.html', status=status)

if __name__ == '__main__':
    app.run()
```
