````python
## Task 1.1
class Tile:
    def __init__(self, suit, value):
        self.__suit = suit
        self.__value = str(value)
    def get_suit(self):
        return self.__suit
    def get_value(self):
        return self.__value
    def __repr__(self):
        return f"({self.__suit},{self.__value})"
        
````

````python
## Task 1.2
class Queue:
    def __init__(self): 
        self.queueList = [] ## Dynamically sized Queue
        self.headPointer = 0
        self.tailPointer = -1
        self.numberItems = 0

    def isEmpty(self):
        return self.numberItems == 0
    
    def enqueue(self, tile):
        self.queueList.append(tile)
        self.numberItems += 1

    def dequeue(self): 
        if self.isEmpty():
            return None
        self.numberItems -= 1
        tile = self.queueList.pop(self.headPointer)
        return (tile.get_suit(), tile.get_value())

    def printQueue(self):
        print("First", end=" ")
        for i in range(self.numberItems):
            print(self.queueList[i], end=" ")
        print("Last")
````

````python
## Task 1.3

import random
class Wall(Queue):
    def __init__(self):
        super().__init__()

    def build_wall(self):

        for suit in ("Bamboo", "Character", "Dot"):
            for value in range(1,10):
                self.queueList.extend([Tile(suit,value) ]*4)

        suit = "Dragon"
        for value in ("Red", "Green", "White"):
            self.queueList.extend( [Tile(suit, value)] * 4)

        suit = "Wind"
        for value in ("East", "South", "West", "North"):
            self.queueList.extend( [Tile(suit, value)] * 4)

        self.numberItems = len(self.queueList)
        
    def shuffle(self):
        random.shuffle(self.queueList)
        
````

````python
## Task 1.3 test case
wall = Wall()
wall.build_wall()
print(wall.numberItems)
````

````python
### Task 1.4

class PlayHand(Queue):
    def __init__(self):
        super().__init__()

    def add_tile(self, tile):
        self.enqueue(tile)

    def get_tiles(self):
        self.printQueue()

    def sort_hand(self):
        for i in range( len(self.queueList)-1):
            swap = False
            for j in range(len(self.queueList)-1-i):
                if self.queueList[j].get_suit() > self.queueList[j+1].get_suit():
                    self.queueList[j], self.queueList[j+1] = self.queueList[j+1], self.queueList[j]
                    swap = True
                elif self.queueList[j].get_suit() == self.queueList[j+1].get_suit():
                    if self.queueList[j].get_value() > self.queueList[j+1].get_value():
                        self.queueList[j], self.queueList[j+1] = self.queueList[j+1], self.queueList[j]
                        swap = True
            if not swap:
                break
````
