```python
## 3.1 is it necessary to implement exception handing ? - No

units = {"KB": 1e3, "MB": 1e6, "GB": 1e9, "TB": 1e12} # metric units
def task3_1(quantity_of_data:str):

    if quantity_of_data[-2:] in units.keys() and quantity_of_data[:-2].isdigit():
        return int(int(quantity_of_data[:-2]) * units[quantity_of_data[-2:]])
    else:
        raise Exception("invalid data") # return "invalid data"
```

```python
#3.2
units2 = {"KiB": 2**10, "MiB": 2**20, "GiB": 2**30, "TiB": 2**40} # binary units
def task3_2(quantity_of_data:str):

    if quantity_of_data[-3:] in units2.keys() and quantity_of_data[:-3].isdigit():
        return int(int(quantity_of_data[:-3]) * units2[quantity_of_data[-3:]])
    else:
        return task3_1(quantity_of_data)

```

##### Using Regular expression

```python
## Custom Exception
class InvalidData(Exception):
    def __init__(self, s):
        super().__init__()
        self.__data=s
    def __str__(self):
        return(f"Invalid data:{self.__data}")
```

```python
import re
metric_units = {"KB": 1e3, "MB": 1e6, "GB": 1e9, "TB": 1e12} # metric units
binary_units = {"KiB": 2**10, "MiB": 2**20, "GiB": 2**30, "TiB": 2**40} # binary units
units = { **metric_units, **binary_units}
def to_bytes(quantities_of_data:str)->int:

    pattern = f"(\d+)({'|'.join(units.keys())})" # (\d+)(KB|MB|GB|...|TiB)
    ma = re.match(pattern, quantities_of_data)
    if ma:
        #print(ma.group(1), ma.group(2))
        return int(ma.group(1))*units[ma.group(2)]
    else:
        raise InvalidData(quantities_of_data)
```

```python
to_bytes("32GiB")
```

```python
# 3.3
def convert(quantity_of_data, target_units):
    try:
        value_bytes = to_bytes(quantity_of_data)
        return value_bytes / units[target_units]
    except KeyError:
        raise InvalidData(target_units)
```

```python

convert("32GiB","GiB")
```

```python
## 4.1
MR = [ ["Empty" for _ in range(6)]for _ in range(10)] #10x6 = 60
TR1 =[ ["Empty" for _ in range(4)]for _ in range(5)] #5x4 = 20
TR2 =[ ["Empty" for _ in range(5)]for _ in range(6)] #6x5 = 30

def display(room:list):
    for row in room:
        for col in row:
            print(f"{col:<8}", end="")
        print()
```

```python
display(MR)
print("-"*80)
display(TR1)
print("-"*80)
display(TR2)
```

```python
## 4.2
def populate(room, candidates):
    try:
        ASC = True
        for col in range(len(room[0])):
            if ASC:
                for row in range(len(room)):
                    room[row][col] = candidates.pop(0) # index error when no more candidates
                ASC = not ASC
            else: # opposite direction
                for row in range(len(room)-1,-1,-1 ):
                    room[row][col] = candidates.pop(0) # index error when no more candidates
                ASC = not ASC
    except IndexError :
        print("No more Candidates to allocate")
```

```python
f = open("CANDIDATES.txt")
candidates = f.read().split("\n")
```

```python
rooms = [MR, TR1, TR2]
names = ["Multipurpose Room", "Tutorial Room 1", "Tutorial Room 2"]
for i,room in enumerate(rooms):
    print(names[i])
    populate(room, candidates)
    display(room)
```

```python
## 1.1
students =  [ line.strip().split(",") for line in open("students.txt") ]
```

```python
## 1.2
while True:
    target = input("Enter student name:")
    for student in students:
        name, age, gender = student
        if target == name:
            print(f"Age: {age} Gender {gender}")
            break
    else:
        print("name not found")
        continue
    break
```

```python
# 1.3
#students =  [ line.strip().split(",") for line in open("students.txt") ]
ages= []
male = 0
female = 0
for student in students:
        name, age, gender = student
        ages.append(int(age))
        if gender == "M":
            male+=1
        else:
            female +=1
print(f"Average age is : {round(sum(ages)/len(ages),1)}")
print(f"Number of males: {male}, Number of females: {female}")
```

```python
with open("updated_students.txt", "w") as f:
    lines=[]
    for student in students:
        name, age, gender = student
        lines.append(f"{name.upper()},{age},{gender}")
    f.write("\n".join(lines))
```
