### Exercise 1 fixed size BST


What is a Binary Search Tree ?
- BST is a subset of a Binary Tree
- Every node in a BST has at most 2 child nodes.
- the left child node's key must be smaller than the parent node
- the right child node's key  must be larger than the parent node


##### What is a fixed size BST ?
    - Uses a fixed size array to store Nodes.
    - Left and Right Pointers in a node are indexes in the array

````python
## Node class
class Node:
    def __init__(self, data=None):
        self.data = data ## Note that data can be a composite structure that consists of (key, value)
        self.left = None #[this is an index]
        self.right = None

    def __repr__(self):
        return f"{self.left}:{self.data}:{self.right}"
````

````python
def AddItemToBinaryTree(NewFreeItem):
    global ThisTree, Root, NextFreePosition

    if NextFreePosition == -1: ## Check for free space
            return False

    if Root == -1: # Empty Tree
        Root = NextFreePosition
    else:
        cur_position = Root
        last_move = "X"
        while True: # traverse till leaf is found

            prev_position = cur_position
            if NewFreeItem < ThisTree[cur_position].data:
                last_move = "L"
                cur_position = ThisTree[cur_position].leftP
            else:
                last_move = "R"
                cur_position = ThisTree[cur_position].rightP

            if cur_position == -1:
                break

        if last_move == "R":
            ThisTree[prev_position].rightP = NextFreePosition
        else:
            ThisTree[prev_position].leftP = NextFreePosition

## insert the free node into the BST and
    insert_position = NextFreePosition #

    NextFreePosition = ThisTree[insert_position].leftP ## this line alone is not enough

    ## create the new node at insert_position
    ThisTree[insert_position].data = NewFreeItem
    ThisTree[insert_position].leftP = -1
    ThisTree[insert_position].rightP = -1


````
