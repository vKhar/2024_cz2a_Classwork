```python
## 2.1
class Node:
    def __init__(self, data=-1, left=-1, right=-1):
        self.__data = data
        self.__left = left
        self.__right = right
    def get_data(self):
        return self.__data
    def get_left(self):
        return self.__left
    def get_right(self):
        return self.__right
    
    def set_data(self, v):
        self.__data = v
    def set_left(self, v):
        self.__left = v
    def set_right(self, v):
        self.__right = v

    def __repr__(self):
        return f"<{self.__left}:{self.__data}:{self.__right}"
```

```python
## 2.2
def calculate_nodes(n):
    return (2**n) -1 #[3m]
```

```python
## 2.2
def calculate_nodes(n):
    if n == 1:
        return "A 1 level number tree has 1 node"
    stack = [1]
    for i in range(1, n):
        stack.append(stack[-1] * 2)
    print(f"A {n} level number tree has {sum(stack)} nodes")
```

```python
## 2.3
print(calculate_nodes(1))
print(calculate_nodes(3))
print(calculate_nodes(5))
print(calculate_nodes(10))
```
