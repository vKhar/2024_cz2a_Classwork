```python
## Terminate Server using "bye"

from socket import socket
s = socket()
s.connect(('127.0.0.1', 1234) )
s.sendall(b"bye") #EOT symbol
ack = s.recv(5)
if ack.decode()== chr(6): # ACK symbol
    print("message acknowledged")
s.close()
```

```python
## Terminate Server using EOT

from socket import socket
s = socket()
s.connect(('127.0.0.1', 1234) )
s.sendall(chr(4).encode()) #EOT symbol
ack = s.recv(5)
if ack.decode()== chr(6): # ACK symbol
    print("message acknowledged")
s.close()
```

```python
## Send a huge chunk in 1 send
from socket import socket
f = open("HAMLET.txt", "rb") #binary read
bytes_stream = f.read() + chr(3).encode() #ETX
f.close()
print("Start of File Transfer")
s = socket()
s.connect( ('127.0.0.1',1234)  )
print(len(bytes_stream))
s.sendall(bytes_stream)
print("End of File Transfer")
```

```python
## Guess Simplify Client Code (Mr Leong)
import socket, time

s = socket.socket()
s.connect(('127.0.0.1', 9999))

correct = False
while not correct :
    raw = s.recv(1024)
    code = raw.decode()
    print(f"Received {code}")

    if code == 'Low':
        print('Your guess is too low.')
    elif code== 'High':
        print('Your guess is too high.')
    elif code == 'GUESS':
        print("Guess a number between 1 to 100")
    elif code == 'Correct':
        print('You guessed correct!')
        correct = True
        break
    else:
        print(f"{code} is not recognised")
    guess = input('Enter guess (1-100): ')
    s.send(guess.encode())

s.close()
```
