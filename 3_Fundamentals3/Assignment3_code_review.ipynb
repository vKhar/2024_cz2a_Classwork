A text file, TIDES. TXT, contains the low and high tide information for a coastal location for each
day of a month. Each line contains tab-delimited data that shows the date, the time, whether the
tide is high or low and the tide height in metres.

Each line is in the format:
```
YYYY-MM-DD\tHH:mm\tTIDE\tHEIGHT\n
```
-   The date is in the form YYYY-MM-DD, for example, 2019-08-03 is 3rd August, 2019
-   The time is in the form HH:mm, for example, 13:47
-   TIDE is either HIGH or LOW
-   HEIGHT is a positive number shown to one decimal place
-   \t represents the tab character
-   \n represents the newline character

The text file is stored in ascending order of date and time

##### Task 2.1
Write Python code to:
-   read the tide data from a text file
-   find the highest high tide and print this value
-   find the lowest low tide and print this value.
Use TIDES.TXT to test your program code.

````python
#YYYY-MM-DD\tHH:mm\tTIDE\tHEIGHT\n
try:
    tide_list = []
    tide_data = open("TIDES.txt")
    for line in tide_data:
        tide_list.append(int(float(line.strip()[-3:]) * 10))
        max_tide = max(tide_list)
        min_tide = min(tide_list)
        low_tide = min_tide / 10
        high_tide = max_tide / 10
    print(high_tide, low_tide)
except Exception as ex:
    print(ex)
````

#### Task 3
The task is to implement a Caesar cypher encryption algorithm.
A Caesar cypher encodes each letter with a different letter. A 10-place Caesar cypher uses the
ASCII value of each letter and adds the number 10 to it. An ASCII value is an integer used to represent a letter or symbol.
For example:

-   The letter 'A' has the ASCII value 65. Adding 10 to the number will give the ASCII value 75.
The character for 75 is 'K'.
-   When an uppercase letter's code goes beyond 'Z' it returns to 'A'. For example, the character
'Z' will be encrypted as 'J'.
-   When a lowercase letter's code goes beyond 'z' it returns to 'a'. For example, the character 'x'
will be encrypted as 'h'.
-   Spaces ' ', are replaced with the character '!'.

You will only need to convert letters and spaces. If the character is invalid, -1 is returned.
The following table shows the ASCII values of some of the characters.

| Character | ASCII value |
| --- | --- |
|  A  | 65  |
|  Z  | 90  |
|  a  | 97  |
|  z  | 122 |
| ' '(space) | 32 |
| ! | 33 |

In ASCII the letters follow on numerically. For example, the letter 'A' is 65, 'B' is 66, 'C' is 67 etc.

You may use the Python function ```ord(letter)``` to return the ASCII value of a letter and the ```chr(ascii_value)``` which will return the letter when given the ASCII value.



##### Task 3.1

Write program code for a function that takes a character as a parameter and returns the ASCII

````python
def encrypt(letter):
    if letter==' ':
        return '!'
    elif ord(letter) in range(65,91) or range(97,123) :
        return ord(letter)+10
    else:
        return -1
````

##### To implement a wrap-around functionality on a range of values $[lb, ub]$ using modulo:
- given an initial value $u$ in $[lb, ub]$, increment $u$ by an amount $i$ such that the new value, $v$, if > ub will wrap back to start at $lb$

- lines 7 to 12 can be re-written with the code shown below

````python
## The generic wrap-around function
def wrap_around(lb,ub,u,i):
  return ((u+i)-lb)%(ub-lb+1) +lb
````

````python
ascii_letter = "z" # example test case
if ascii_letter.isupper():
    encrypted_ascii = wrap_around( ord("A"), ord("Z"), ord(ascii_letter), 10)
elif ascii_letter.islower():
    encrypted_ascii = wrap_around( ord("a"), ord("z"), ord(ascii_letter), 10)
print(chr(encrypted_ascii))
````
