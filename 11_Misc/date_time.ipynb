### Why and when do you need to use the datetime module
-   comparing dates when sorting
-   calculating number of days between 2 dates

```python
from datetime import datetime
```

```python
day_1 = "030125" ## string compare
day_2 = "121225"
```

```python
day_1 = "3/1/25" ## string compare
day_2 = "12/12/25"
day_1 < day_2
```

```python
yi_fan_wedding_date = datetime.now() + timedelta( 10 * 365) ## add number of days to now
print(yi_fan_wedding_date.strftime('%d-%m-%Y')) # print datetime object as a str
```

```python
print(f" I am {datetime.now() - datetime.strptime('13*5*2007','%d*%m*%Y')} days old") # datetime substration
```

```python
from  datetime import datetime, timedelta
day_1 = datetime.strptime("3/1/25", "%d/%m/%y")  ### %d %m %y %Y %H %M %S
day_2 = datetime.strptime("12/1/26", "%d/%m/%y")
day_diff = day_2 - day_1 #substration
print(day_diff.days)
```
