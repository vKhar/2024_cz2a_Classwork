### Why and when do we use hexadecimal System ?
- Humans use Denary or Decimal System (Base 10)
- Computers use Binary System (Base 2)

***Hexadecimal system*** is an intermediary or bridging system to provide a more ***human-friendly*** way to represent binary (base-2) numbers. It acts as a shorthand by grouping binary digits(4 bits) into manageable chunks (a hexadecimal digit), making it easier for programmers and engineers to read and interpret the data.

For example, A mac address(48 bit) for a device on the network is identifies as :

- ```111101001010100000001101101000111100111110001010```
- It can be grouped into chunks of 4-bits as follows:
    ```
     1111 0100 1010 1000 0000 1101 1010 0011 1100 1111 1000 1010
    ```
- each 4-bit will then map into a single hexadecimal digit as follows:
    ```
    F 4 A 8 0 D A 3 C F 8 A
    ```
    ![](Bin-Hex.jpg)
- we can further group 2 hexadecimal digits together because they represent 8 bits or 1 byte. Computing devices uses 1 byte as the smallest unit of for data storage and manipulations.
    ```
    F4-A8-0D-A3-CF-8A
    ```

````python
def hex2dec(hex_str:str)->int:
    conversion_str = "0123456789ABCDEF"
    place_value = len(hex_str) - 1 # Most sig place value
    total = 0 #will count the denary value

    for hex_digit in hex_str: # for i in range(len), 0, 1, ..
        den_value = conversion_str.index(hex_digit.upper())
        total += den_value * (16**place_value) ## place_value (n-1)-i
        place_value-=1

    return total

````

````python
def base_k2dec(base_str:str, k: int)->int:
    conversion_str = "0123456789ABCDEF"
    place_value = len(base_str) - 1 # Most sig place value
    total = 0 #will count the denary value

    for hex_digit in base_str: # for i in range(len), 0, 1, ..
        den_value = conversion_str.index(hex_digit.upper())
        total += den_value * (k ** place_value) ## place_value (n-1)-i
        place_value-=1

    return total

````

````python
base_k2dec("1042301",5), int("1042301", 5)
````

````python
def dec2base_k(dec_val:int, k:int)->str:
    conversion_str = "0123456789ABCDEF"
    hex_str = ""
    while dec_val:
        remainder = dec_val % k
        hex_value = conversion_str[remainder]
        hex_str = hex_value + hex_str #it's read backwards
        dec_val //= k

    return hex_str
````

````python
dec2base_k(1234, 16), hex(1234)
````

````python
def bin2dec(bin_str:str)->int:
    total = 0
    place_val = len(bin_str) - 1
    for i in bin_str:
        total += int(i)*(2**place_val)
        place_val -= 1

    return total
````

````python
def dec2bin(dec_val:int)->str:
    bin_str = ""
    while dec_val:
        remainder = dec_val % 2
        bin_str = str(remainder) + bin_str
        dec_val //= 2
    return bin_str
````

````python
def hex2bin(hex_str:str)->str:
    return dec2bin(hex2dec(hex_str)) #either this or hard coded dictionary
````

````python
def bin2hex(bin_str:str)->str:
    return dec2hex(bin2dec(bin_str)) #either this or split bin_str into groups of 4 and then the reverse of the dict above
    #Note, just append the leading zeros into the str if the length is not multiple of 4 for the other approach (for convenience)
````
