```python
dates = [
    "2024_2_10",
    "2024_11_10",
    "2024_12_2",
    "2024_12_17"
] ## yyyy-m-d
sorted(dates)
```

```python
## Using datetime.strpdate() to convert str to datetime object
from datetime import datetime
dates = [ datetime.strptime(s, "%Y_%m_%d") for s in dates]
sorted(dates)
```
