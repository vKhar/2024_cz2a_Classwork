#### Encoding using UTF-8, UTF-16, UTF-32

Every code points ( from 0 to 1114111) can be encoded in either UTF-8, UTF-16, UTF-32 format
- UTF-8 is variable length, 1 byte, 2 bytes , 3 bytes or 4 bytes
- UTF-16 is variable lenhth, 2 bytes or 4 bytes
- UTF-32 is fixed length 4 bytes

#### Why do we need up to 4 bytes to encode all the code points instead of 3 bytes ?
-   The highest code point is `1114111 (U+10ffff)`, if we convert this to binary value, it will just need 21 bits

-   There are overhead bits used in the encoding:

    - 1 byte: 0xxxxxxx (7 bits)
    - 2 bytes: 110xxxxx 10xxxxxx (11 bits)
    - 3 bytes: 1110xxxx 10xxxxxx 10xxxxxx (16 bits)
    - 4 bytes: 11110xxx 10xxxxxx 10xxxxxx 10xxxxxx (21 bits)

```python
## helper function to print bytes in binary
## Python does not have a built-in function to print bits
## use -le, or -be for utf-16-be, utf-32-be
def print_bin(symbol)->str:
    bytes_array = symbol.encode("utf-8")
    print( ":".join (f"{i:^8d}" for i in range(len(bytes_array) ) ))
    print( ":".join ( f"{b:08b}" for b in bytes_array ) )
```

```python
print(f"{'symbol':^8}{'dec':^8}{'hex':^8}")
print(f"{symbol:^8}{ord(symbol):^8}{hex(ord(symbol)):^8}")
print("Encoded in memory as bytes array:")
print_bin(symbol)
```
