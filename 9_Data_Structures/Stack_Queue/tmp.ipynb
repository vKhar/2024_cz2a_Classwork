A stack is used to store characters.

### Task 1
Write program code to implement the stack and the operations specified.
Your code should allow operations to:
- push an item on to the stack
- pop an item off the stack
- determine the size of the stack. A size of zero indicates that the stack is empty.

```python
class CharStack:
    """Minimal stack storing single-character entries."""

    def __init__(self) -> None:
        self._data: list[str] = []

    def push(self, item: str) -> None:
        if len(item) != 1:
            raise ValueError("Only single characters can be pushed onto the stack.")
        self._data.append(item)

    def pop(self) -> str:
        if not self._data:
            raise IndexError("Cannot pop from an empty stack.")
        return self._data.pop()

    def size(self) -> int:
        return len(self._data)

    def is_empty(self) -> bool:
        return not self._data

    def peek(self) -> str:
        if not self._data:
            raise IndexError("Cannot peek into an empty stack.")
        return self._data[-1]


# Demo usage showing the supported operations.
stack = CharStack()
stack.push("A")
stack.push("B")
stack.push("C")
print("Top of stack:", stack.peek())
print("Popped value:", stack.pop())
print("Current size:", stack.size())
print("Is empty?:", stack.is_empty())
```
