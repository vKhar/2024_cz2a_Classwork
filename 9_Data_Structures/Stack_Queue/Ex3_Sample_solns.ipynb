```python
def ValidateUserID(id):
    # length, presence check
    if len(id) != 9:
        return 1
    # format check
    try:
        prefix, suffix = id.split("_")
    except ValueError:
        return 4
    
    if prefix != '2015':
        return 2
    
    if not suffix.isdigit():
        return 3
    return 0
    # range check
```

```python
print(ValidateUserID("2015_9"))
print(ValidateUserID("2014_9999"))
print(ValidateUserID("2015_p000"))
```

```python
## Fized Size Circular
class Queue:
    def __init__(self, size):
        self.array = [None for _ in range(size)]
        self.in_use = 0
        self.front = 0 # delete position
        self.back = 0 # insert position

    def is_empty(self):
        return self.in_use == 0
    
    def is_full(self):
        return self.in_use == len(self.array)

    def enqueue(self, data):
        if self.is_full():
            print("Full")
            return False
        else:
            self.array[self.back] = data
            self.back = (self.back+1)%len(self.array)
            self.in_use+=1
            return True

    def dequeue(self):
        if self.is_empty():
            print("Empty")
            return None
        else:
            ret = self.array[self.front]
            self.array[self.front] = None
            self.front = (self.front + 1)%len(self.array)
            self.in_use -=1
            return ret
    

    def __repr__(self):
        return f"{self.array}"
    
class PrintJob():
    def __init__(self, user_id, terminal_id, file_size):
        self.__user_id = user_id
        self.__terminal_id = terminal_id
        self.__file_size = file_size
    def __repr__(self):
        return f"{self.__user_id}:{self.__terminal_id}:{self.__file_size}"

class PrintQueue(Queue):
    def __init__(self, size, name):
        super().__init__(size)
        self.__name = name
    
    def show(self):
        ## Iterate from head to tail
        if self.is_empty():
            print("Empty Queue")
        else:
            
            cur = self.front
            count = 0
            while count < self.in_use:
                print(f"{self.array[cur]}")
                cur = (cur+1)%(len(self.array))
                count+=1
```

```python
menu = '''
| 1. | New print job added to print queue |
| 2. | Next print job output from printer |
| 3. | Current print queue displayed      |
| 4. | End                                |
'''
```

```python
njc_printer =PrintQueue(2,"njc_print")
while True:
    print(menu)
    try:
        choice = int(input("Enter 1 or 2 or 3 or 4"))
        if choice not in (1,2,3,4):
            raise ValueError
    except ValueError:
        print("Invalid input")
        continue
    if choice == 4:
        break
    if choice == 1:
        new_job = PrintJob(
            input("Enter User ID"),
            input("Enter Terminal ID"),
            input("file size in KB")
        )
        if njc_printer.enqueue(new_job):
            print("Job submitted")
        else:
            print("Falied to submit print job")
    if choice == 2:
        print_job = njc_printer.dequeue()
        if print_job:
            print(f''' Printing {print_job}''')
        else:
            print("Nothing to print")
    if choice == 3:
        #print(njc_printer)
        njc_printer.show()
```
