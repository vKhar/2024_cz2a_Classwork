```python
## Sentinel node implementation

class LinkedList: # Node class
    def __init__(self, data=None):
        self.__data = data
        self.__next = None

    def get_data(self):
        return self.__data
    def set_data(self,data):
        self.__data = data
    def get_next(self):
        return self.__next
    def set_next(self, next):
        self.__next = next


    def insert(self, n):
        new_LL = LinkedList(n)
        if self.get_next() == None:
            self.set_next(new_LL)
        else:
            new_LL.set_next (self.get_next())
            self.set_next(new_LL)

    def delete(self,n):
        if self.get_next() == None:
            return None
        else:
            prev = self # sentinel node
            cur = self.get_next()
            while cur and cur.get_data() != n:
                prev = cur
                cur = cur.get_next()
            if cur : # node found
                ret = cur.get_data()
                prev.set_next (cur.get_next())
                return ret
            else: ## not found
                return None

    def search(self, n):
        if self.get_next() == None:
            return False
        else:
            cur = self.get_next()
            while cur:
                if cur.get_data() == n:
                    return True
                cur = cur.get_next()
            return False
    def count(self):
        ret=0
        if self.get_next() == None:
            return ret
        else:
            cur = self.get_next()
            while cur:
                ret+=1
                cur = cur.get_next()
        return ret

    def to_String(self):
        ret = []
        if self.get_next() == None:
            return f"{ret}"
        else:
            cur = self.get_next()
            while cur != None:
                ret.append(cur.get_data())
                cur = cur.get_next()
            return f"{ret}"
```

```python
ll = LinkedList()
## Test cases
for i in open("Task3data.txt").read().split("\n"):
    ll.insert(int(i))
print(ll.to_String())

print(ll.delete(7))
print(ll.to_String())
ll.search(2)
```

```python
class SortedLinkedList(LinkedList):
    def __init__(self, data= None):
        super().__init__(data)

    def insert(self, n):
        new_LL = SortedLinkedList(n)
        if self.get_next() == None:
            self.set_next(new_LL)
        else:
            prev = self
            cur = self.get_next()
            while cur and n > cur.get_data():
                prev = cur
                cur = cur.get_next()
            new_LL.set_next (cur)
            prev.set_next(new_LL)

```
